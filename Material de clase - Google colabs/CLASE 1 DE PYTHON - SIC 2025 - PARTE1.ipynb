**INDICES EN LISTAS**

Cada elemento de una lista tiene una posición (número entero) llamada **índice.**

* Los índices en Python empiezan desde 0.

* También se pueden usar índices negativos para acceder desde el final.

```python
# EJEMPLO DE USO DE INDICES EN LISTAS

numeros = [10, 20, 30, 40, 50]

print(numeros[0])   # 10 (primer elemento)
print(numeros[2])   # 30 (tercer elemento)
print(numeros[-1])  # 50 (último elemento)
print(numeros[-2])  # 40 (penúltimo elemento)
```

**Diferencia entre índice e indexación**

>**Índice:** el número que representa la posición de un elemento en la lista.
Ejemplo: en [10, 20, 30, 40, 50], el número 20 está en el índice 1.

>**Indexación:** la operación de acceder a un elemento usando su índice.
Ejemplo: mi_lista[1] es la indexación que devuelve el valor 20.

**En otras palabras:**

* El índice es el número de la posición.

* La indexación es la acción de usar el índice para obtener un valor.

**CALCULO DE LA MEDIA DE LA EDAD UTILIZANDO LISTAS**

```python
person1 = ['David Doe', 20, 1, 180.0, 100.0]
person2 = ['John Smith', 25, 1, 170.0, 70.0]
person3 = ['Jane Carter', 22, 0, 169.0, 60.0]
person4 = ['Peter Kelly', 40, 1, 150.0, 50.0]

person_list = person1 + person2 + person3 + person4
n_persons = int (len(person_list) / 5)
age_sum = 0.0

for age in person_list[1 :: 5] :
  age_sum += age
  average_age = float(age_sum) / n_persons
print('la edad media es ' + str(average_age) + '.')
```

**USO DE INDICE NEGATIVO EN LISTAS**

**QUE ES EL INDICE NEGATIVO EN UNA LISTA?**

En Python, los índices negativos permiten acceder a los elementos de una lista desde el final hacia el inicio.

* El índice -1 corresponde al último elemento.

* El índice -2 corresponde al penúltimo elemento, y así sucesivamente.

**SINTAXIS BASICA**

```python
lista = [10, 20, 30, 40, 50]

print(lista[-1])  # Último → 50
print(lista[-2])  # Penúltimo → 40
print(lista[-3])  # Antepenúltimo → 30
```

```python
frutas = ["manzana", "pera", "naranja", "uva", "sandía"]

print("Última fruta:", frutas[-1])     # sandía
print("Penúltima fruta:", frutas[-2])  # uva
```

```python
# Simulación de registros de log de un sistema
logs = [
    "2025-08-24 10:00 - INFO - Inicio del sistema",
    "2025-08-24 10:05 - WARN - Memoria alta",
    "2025-08-24 10:07 - ERROR - Conexión perdida con DB",
    "2025-08-24 10:08 - INFO - Reconexión establecida",
    "2025-08-24 10:10 - INFO - Proceso completado"
]

# Ver los últimos 2 registros
ultimos_logs = logs[-2:]
print("Últimos registros importantes:")
for log in ultimos_logs:
    print(log)
```

**¿Por qué se inventó esto?**

Porque es muy común querer acceder a los últimos elementos de una lista sin tener que calcular su posición.

```python
#Obtencion de los utimos 3 elementos de una lista usando indices negativos y slicing.

lista = [1, 2, 3, 4, 5]
print(lista[-3:])  # últimos 3 elementos → [3, 4, 5]
```

**ORDENAMIENTO DE LISTAS USANDO EL METODO SORT**

El método **sort()** en Python es una función propia de las listas que permite ordenar los elementos en orden ascendente por defecto.

**Funcionamiento básico**

>Cuando tienes una lista con valores desordenados, puedes aplicar lista.sort() para que sus elementos se acomoden de menor a mayor.

**Características principales:**

*   Orden ascendente por defecto: Ordena los elementos de menor a mayor.
*   Orden descendente: Puedes agregar el argumento reverse=True para invertir el orden.

```python
# Ordenación ascendente de los elementos de la lista1
list1 = [20, 10, 40, 50, 30]
list1.sort()
print(list1)

# Ordenación descendente de los elementos de la lista1
list2 = [20, 10, 40, 50, 30,0]
list2.sort(reverse = True)
print(list2)

```

**EJEMPLO DE ORDENAMIENTO SIN EL USO DE SORT**

```python
# Lista original
lista = [20, 10, 40, 50, 30]

# Orden ascendente con ciclos (Burbuja)
for i in range(len(lista)):
    for j in range(len(lista) - 1):
        if lista[j] > lista[j + 1]:
            # Intercambiamos los elementos
            lista[j], lista[j + 1] = lista[j + 1], lista[j]

print("Orden ascendente:", lista)

# Otra lista para descendente
lista2 = [20, 10, 40, 50, 30]

# Orden descendente con ciclos (Burbuja)
for i in range(len(lista2)):
    for j in range(len(lista2) - 1):
        if lista2[j] < lista2[j + 1]:
            lista2[j], lista2[j + 1] = lista2[j + 1], lista2[j]

print("Orden descendente:", lista2)
```

**Ordenamiento personalizado:**

>Con el parámetro key puedes definir cómo se comparan los elementos.

Ejemplo 1: ordenar palabras por su longitud.

```python
palabras = ["perro", "gato", "elefante", "oso"]
palabras.sort(key=len)
print(palabras)
# Resultado: ['oso', 'gato', 'perro', 'elefante']
```

Ejemplo 2: Ordenar las palabras por su primera letra

```python
palabras = ["uva", "pera", "manzana", "banana"]

# Ordenamos según la primera letra
palabras.sort(key=lambda palabra: palabra[0])

print(palabras)
```

# **REPASO USO DE APPEND, REMOVE, POP, DEL Y REBANADO**

**.append()**

*   Es un método que agrega un elemento al final de la lista.

*   Funciona para ampliar una lista con nuevos valores sin necesidad de crear otra.
*   Lo usarías cuando quieras ir construyendo una lista de forma dinámica, por ejemplo, cuando vas registrando datos en un programa.







```python
frutas = ["manzana", "pera"]
frutas.append("naranja")
print(frutas)
```

**.remove()**

*   Es un método que elimina la primera coincidencia de un valor dentro de la lista.

*   Funciona cuando conoces el valor pero no el índice.
*   Lo usarías, por ejemplo, para borrar un elemento específico de un registro o lista de tareas.











```python
numeros = [1, 2, 3, 2, 4, 2, 2]
numeros.remove(2)
print(numeros)
```

**.pop()**

*   Es un método que elimina un elemento por índice y además lo devuelve.
*   Funciona cuando quieres quitar algo de la lista pero también necesitas usar ese valor después.
*   Lo usarías, por ejemplo, para extraer el último elemento introducido o para implementar pilas/colas.
*   A diferencia de remove(), que elimina por valor, pop() elimina por índice.
*   Si no se le pasa un índice, por defecto elimina el último elemento.

Este metodo guarda el ultimo valor por si en algun caso deseamos hacer algo con el mismo, como pilas o listas de tareas.











```python
colores = ["rojo", "verde", "azul"]
ultimo = colores.pop()
print(colores, "Eliminado:", ultimo)
```

```python
cola = ["cliente1", "cliente2", "cliente3"]

atendido = cola.pop(0)  # siempre quitamos el primero, indice 0
print("Atendido:", atendido)
print("Cola ahora:", cola)
```

**Rebanado**

*   Es una técnica que permite obtener una parte de la lista usando índices de inicio y fin.

*   Funciona para crear sublistas sin modificar la original.
*   Lo usarías cuando necesites acceder a un rango de datos, por ejemplo, tomar los tres primeros elementos.











```python
nombres = ["Ana", "Luis", "Karla", "Pedro", "Sofía"]
print(nombres[1:4])
```

# **QUE SON LISTAS Y QUE SON TUPLAS?**

**¿Qué es una lista?**

>Una lista en Python es una colección ordenada y mutable de elementos.


*   Ordenada: cada elemento tiene una posición (índice).
*   Mutable: se puede modificar (agregar, eliminar o cambiar valores).


Puede contener distintos tipos de datos (números, cadenas, otras listas, etc.).

```python
lista = [10, "hola", 3.5]
lista[1] = "adiós"   # se puede cambiar
print(lista)
```

**¿Qué es una tupla?**

>Una tupla es una colección ordenada e inmutable de elementos.

*   Ordenada: al igual que la lista, mantiene un índice.
*   Inmutable: no se puede cambiar después de creada (no admite agregar, quitar ni modificar elementos).

Se usa cuando los datos no deben alterarse.

```python
colors = ("red", "green", "blue")
colors
```

```python
#Modificacion de tuplas
tupla = (10, 20, 30)

# Intentamos cambiar el segundo elemento
tupla[1] = 50
print(tupla)
```

# **USO DE DICCIONARIOS - INTRODUCCION**

image.png

>Un diccionario es una estructura de datos en Python que almacena pares de clave:valor.

* A diferencia de las listas o tuplas, en un diccionario no accedemos por índice numérico, sino por la clave.

* Las claves son únicas y no se pueden repetir.

* Los valores pueden repetirse y pueden ser de cualquier tipo (números, cadenas, listas, otros diccionarios, etc.).

**SINTAXIS BASICA.**

> nombre_diccionario = {clave: valor, clave2: valor2, ...}


```python
diccionario = {
    "nombre": "Cleidys",
    "edad": 21,
    "ciudad": "Panama"
}

print(diccionario)
```

**Operaciones comunes con diccionarios**

```python
#Insertar un nuevo elemento

diccionario["pais"] = "Argentina"
print(diccionario)
```

```python
#Insertar un nuevo elemento
diccionario["componentes"] = "teclado"
print(diccionario)
```

```python
#Modificar un valor existente

diccionario["Componentes"] = "microfono"
print(diccionario)
```

```python
#Eliminar un elemento

del diccionario["Componentes"]
print(diccionario)
```

**Funciones y operadores utiles**

*   in → Comprueba si una clave existe en el diccionario.
*   not in → Comprueba si una clave no existe.
*   == → Compara si dos diccionarios son iguales (claves y valores idénticos).

*  ' > o < ' → Comparan basados en número de claves, pero no suelen usarse en práctica.





```python
persona = {"nombre": "Ana", "edad": 25, "ciudad": "Madrid"}

print("nombre" in persona)   # True (la clave existe)
print("pais" in persona)     # False (la clave no existe)
```

```python
persona["pais"] = "panama"
print(persona)

print("nombre" in persona)   # True (la clave existe)
print("pais" in persona)     # False (la clave no existe)
```

```python
persona = {"nombre": "Ana", "edad": 25, "ciudad": "Madrid"}

print("pais" not in persona)     # True (no existe esa clave)
print("edad" not in persona)     # False (sí existe esa clave)
```

```python
dic1 = {"a": 1, "b": 2}
dic2 = {"b": 2, "a": 1}
dic3 = {"a": 1, "b": 3}

print(len(dic1)) == len((dic3))  # True (tienen las mismas claves y valores, aunque en distinto orden)
```

```python
dic1 = {"a": 1, "b": 2}
dic2 = {"a": 1, "b": 2, "c": 3}

print(len(dic1) < len(dic2))   # True (dic2 tiene más claves)
print(len(dic1) > len(dic2))   # False
```

# **Formato XML y HTML**

##**¿Qué es JSON?**

JSON = JavaScript Object Notation.

* Es un formato de texto universal para guardar e intercambiar datos.

* Parece un diccionario de Python, pero su gran ventaja es que no depende de un lenguaje de programación.

→ Un servidor en Java, un celular en Kotlin o una página web en JavaScript pueden entender JSON sin problema.

>**¿Por qué se usa JSON?**

* Es más ligero y fácil de leer que otros formatos.

* Es el estándar de comunicación en la web (APIs, aplicaciones móviles, servidores).

* Perfecto para enviar datos entre programas diferentes.

```python
{
  "nombre": "David",
  "apellido": "Doe",
  "edad": 25
}
```

##**¿Qué es XML?**

XML = Lenguaje de Marcado Extensible.

* También sirve para guardar e intercambiar datos.

* Su estructura usa etiquetas de apertura y cierre (como HTML).

```python
<persona>
  <nombre>David</nombre>
  <apellido>Doe</apellido>
  <edad>25</edad>
</persona>
```

> **¿Por qué existe todavía XML?**

* Fue muy popular antes que JSON.

* Se usa en sistemas antiguos y en áreas donde se necesita validar fuertemente los datos (ej. bancos, aerolíneas, documentos oficiales).

Aún lo usan estándares como RSS, SOAP y configuraciones en muchos sistemas.

* Hoy en día, JSON domina la web, pero XML sigue vivo en empresas y sistemas grandes.

## **COMO CONVERTIR DICCIONARIOS A FORMATO JSON USANDO 'dump' y 'loads'?**

**json.dumps() → Diccionario ➝ JSON (en memoria, cadena de texto)**

```python
#Convierte un diccionario en una cadena JSON.
#Se usa cuando quieres ver o enviar datos como texto, pero no guardarlos en archivo.

import json

persona = {"nombre": "David", "edad": 25}

# Convertimos el diccionario a JSON (string)
persona_json = json.dumps(persona, indent=4)

print(persona_json)
```

**json.dump() → Diccionario ➝ JSON (directo en archivo)**

```python
import json

persona = {"nombre": "David", "edad": 25}

# Guardamos el diccionario en un archivo JSON
with open("persona.json", "w") as archivo:
    json.dump(persona, archivo, indent=4)
```

**json.load() → JSON (archivo) ➝ Diccionario**

```python
#Convierte el contenido de un archivo JSON en un diccionario de Python.
#Se usa cuando lees un archivo guardado previamente.

import json

# Abrimos el archivo JSON y lo leemos como diccionario
with open("persona.json", "r") as archivo:
    persona = json.load(archivo)

print(persona["edad"])
```

**Diferencias entre cada funcion**

| Función   | Qué hace                               | Se usa con…          | Caso práctico                                 |
| --------- | -------------------------------------- | -------------------- | --------------------------------------------- |
| `dumps()` | Diccionario → JSON (string en memoria) | Variables en memoria | Enviar datos por internet, mostrar en consola |
| `dump()`  | Diccionario → JSON (archivo)           | Archivos             | Guardar datos en un `.json`                   |
| `loads()` | JSON (string) → Diccionario            | Texto JSON recibido  | Procesar respuesta de una API                 |
| `load()`  | JSON (archivo) → Diccionario           | Archivos             | Leer archivo `.json` guardado                 |


**De JSON a Diccionarios**

```python
import json

# Tenemos un JSON en formato de texto (string)
persona_json = '{"nombre": "David", "edad": 25, "ciudad": "Panamá"}'

# Convertimos ese JSON a un diccionario
persona = json.loads(persona_json)

print(persona)          # {'nombre': 'David', 'edad': 25, 'ciudad': 'Panamá'}
print(persona["nombre"])  # Accedemos como diccionario normal: David
```

**Flujo completo**

```python
#Diccionario a JSON
import json

persona = {"nombre": "Ana", "edad": 28, "ciudad": "Madrid"}

# Guardamos en un archivo JSON
with open("persona.json", "w") as archivo:
    json.dump(persona, archivo, indent=4)
```

```python
#JSON a Diccionario

# Leemos el archivo JSON y lo convertimos en diccionario
with open("persona.json", "r") as archivo:
    persona_cargada = json.load(archivo)

print(persona_cargada)          # {'nombre': 'Ana', 'edad': 28, 'ciudad': 'Madrid'}
print(persona_cargada["edad"])  # 28
```

En resumen:

* Para pasar de diccionario → JSON: usamos dumps() o dump().

* Para pasar de JSON → diccionario: usamos loads() o load().

Así es como podemos enviar datos a otros sistemas y luego recuperarlos en Python.

# **FORMATO PARA UNA MEJOR IMPRESION - Uso de la funcion .format()**

**¿Qué es el formato de impresión?**

>Cuando imprimimos valores en Python con print(), a veces los resultados quedan desordenados o poco claros.

El método .format() permite insertar valores dentro de una cadena de texto de forma ordenada y con control del formato (números, espacios, decimales, etc.).



**¿Por qué usar .format()?**

* Hace el código más claro y ordenado.

* Permite controlar el orden y el formato de los datos.

* Útil para imprimir reportes, tablas o salidas organizadas.

* Funciona con números, texto y otros tipos de datos.



---



**Sintaxis basica**

```python
"texto {marcador}".format(valor)
```

texto {} → el marcador {} indica dónde se va a insertar el valor.

.format(valor) → aquí se pasan los valores que van a reemplazar los {} en orden.


```python
print("Hola {}".format("Python"))
# Hola Python
```



---



**Índices de posición**

Puedes indicar qué argumento quieres usar dentro de los {} mediante índices:

* {0} → primer argumento.

* {1} → segundo argumento.

```python
print("Me gusta {0} , me gusta {1}, y eliminemos {2}".format("Python", "Java", "Fortran"))

print("Me guta {0} , me gusta {3} , no me cae bien {1}, tolero {2}".format("Python", "Java", "C#", "C++"))
```

Varios valores

>Si la cadena tiene varios {}, se reemplazan en orden con los argumentos de format().



```python
print("No me gusta {} pero si {} , tolero {}".format("Java", "Python", "C"))
```

**Formato numérico**

> Con .format() también puedes dar formato a los números:

* {0:4d} → imprime un número entero con al menos 4 espacios.

* {0:05d} → imprime un número entero con al menos 5 dígitos, rellenando con ceros a la izquierda.

```python
print("{0:4d}, {1:5d}, {2:6d}".format(123, 1234555555, 123456789))
```

## **Escritura de archivos y modos**

Diversos modos de entrada/salida de ficheros de la funcion open

La función open abre y trae el archivo 'hola.txt'.
* Importación especificando el modo de apertura de archivos "w" (sólo escritura)
* La función Open devuelve un objeto de archivo.
* El objeto de archivo f utiliza el método write para escribir el texto especificado en un archivo.
* El método f.close cierra el archivo


```python
f = open('hello.txt', 'w')
f.write('hello world!')
f.close()
```

```python
# Modo escritura/lectura de texto
f = open('hello.txt','w+t')

f = open('hello.txt','a+t') # Añadir texto

f.write('Hello World!')

# Escribir en el archivo hello.txt
f.write('Love of my life, you\'ve hurt me\n')
f.write('You\'ve broken my heart and now you leave me\n')
f.write('Love of my life, can\'t you see?\n')
```

Metodo Readline

```python
#Lee dos líneas del fichero hola.txt utilizando el método readline e imprímelas.

f = open('hello.txt','r') #Abrir archivo
s = f.readline() # Lee la primera línea del archivo 'AAA'
print(s, end = " ") # imprimir la línea
s = f.readline() # Leer sequnda línea del fichero 'BBB'
print(s, end = " ") # imprimir la línea
f.close() # Cerrar archivo
```

Usando la declaracion as

```python
with open('hello.txt', 'w') as f:  # abrir y cerrar automáticamente el archivo
  f.write('Hello World!')          # escribir en el archivo hello.txt
```
