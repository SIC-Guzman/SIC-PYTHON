##**UNIDAD 1: Álgebra**

El **álgebra** usa letras (variables) para representar valores que pueden cambiar.
Por ejemplo, si **y = 2x + 3**, entonces para **x = 4**, **y = 11**.

Las ecuaciones describen relaciones entre variables:

- **Primer grado** → líneas rectas
- **Segundo grado** → parábolas



#### **Ejemplo en Python**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y1 = 2*x + 3
y2 = x**2 - 4*x + 2

plt.figure(figsize=(8,5))
plt.plot(x, y1, label="Lineal: y=2x+3")
plt.plot(x, y2, label="Cuadrática: y=x²-4x+2")
plt.legend()
plt.title("Comparación entre ecuación lineal y cuadrática")
plt.grid(True)
plt.show()
```

#### **Explicación**

- La **línea recta** representa un cambio constante.
- La **parábola** crece más rápido porque tiene el término **x²**.
- En **IA**, las ecuaciones ayudan a modelar relaciones entre variables, como peso y altura, edad y salario, etc.

##**UNIDAD 2: Secuencias y Funciones**

#### **Secuencia**

Una **secuencia** es una lista de números que siguen una regla.

Ejemplo (regla: sumar 3 cada vez): 5, 8, 11, 14, 17, ...

#### **Ejemplo en Python - Secuencia**

```python
secuencia = [5 + 3*i for i in range(5)]
print(secuencia)  # [5, 8, 11, 14, 17]
```

##**Funciones**

Una **función** transforma una entrada en una salida.

Ejemplo: **f(x) = x² + 2x + 1**

#### **Ejemplo en Python - Función**

```python
def f(x):
    return x**2 + 2*x + 1

print(f(3))  # 16

# Ejercicio: calcula f(0), f(2), f(-3)
print('f(0)=', f(0))
print('f(2)=', f(2))
print('f(-3)=', f(-3))
```

##**UNIDAD 3: Función Sigmoide**

#### **La Función Sigmoide**

La **funcion sigmoide** es una función matemática con forma de “S” (de ahí su nombre). Su fórmula es:

    f(x) = 1 / (1 + e^{-x})








image.png

Esto significa que toma cualquier número (positivo o negativo) y lo transforma en un valor entre 0 y 1.


| x  | f(x) (sigmoide) |
| -- | --------------- |
| -5 | 0.0067          |
| 0  | 0.5             |
| 5  | 0.993           |


**Podemos ver que:**

Cuando x es muy negativo, la salida se acerca a 0.

Cuando x es muy positivo, la salida se acerca a 1.

En el centro (x=0), la salida es 0.5, el punto medio.

#### **Ejemplo en Python - Gráfica Sigmoide**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 200)
sigmoid = 1 / (1 + np.exp(-x))

plt.figure(figsize=(8,4))
plt.plot(x, sigmoid)
plt.title("Función Sigmoide")
plt.grid(True)
plt.show()
```

**Como funciona?**

Podemos imaginar a la sigmoide como una puerta suave de decisión.

No dice simplemente “sí o no” (como una compuerta digital), sino que decide con grados de certeza:

* Si algo es muy evidente → la salida es cercana a 1 (por ejemplo, 0.98).

* Si algo claramente no lo es → la salida es cercana a 0 (por ejemplo, 0.02).

* Si no está claro → da un valor intermedio (0.4 o 0.6).

**Donde sale en el tema de IA?**

En redes neuronales, la función sigmoide se usa como función de activación.
Cada neurona toma varios valores de entrada, los combina (con una suma ponderada) y aplica la función sigmoide para decidir si “se activa” o no.

Por ejemplo, si una neurona evalúa si una imagen contiene un gato:

Si la combinación de características (orejas, ojos, color) da una

> suma grande → sigmoide ≈ 1 → “sí, hay un gato”.


> Si da una suma pequeña → sigmoide ≈ 0 → “no hay gato”.

Esto convierte salidas continuas en probabilidades.

Por eso, la sigmoide se usa en la última capa de las redes neuronales de clasificación binaria, como modelos que responden sí/no, verdadero/falso, spam/no spam, etc.

#### **Ejemplo en Python - Neurona simple usando sigmoide**

La sigmoide se usa como **función de activación** en redes neuronales. Una neurona toma entradas, las multiplica por sus pesos y luego aplica la sigmoide.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

entradas = np.array([0.2, 0.4, 0.8])
pesos = np.array([0.5, 0.7, 0.3])
suma = np.dot(entradas, pesos)
salida = sigmoid(suma)
print("Suma:", suma)
print("Salida de la neurona:", salida)
```

##**UNIDAD 4: Vectores y Matrices**

#### **Vectores**

Un **vector** es una lista de números que representa dirección, magnitud o características en un espacio. Ejemplo: **v = (2, 3)**

> **En IA**, esos números describen características o atributos de un dato.

**Por ejemplo:**

* Una imagen puede representarse como un vector con miles de valores (cada píxel).

* Una palabra puede representarse con un vector que capture su significado.

* Un perfil de usuario puede representarse con sus preferencias en un vector.

image.png

#### **Ejemplo en Python - Producto punto**

```python
import numpy as np
a = np.array([2, 3])
b = np.array([1, 4])
dot = np.dot(a, b)
print("Producto punto:", dot)
```

**Cómo se usan en IA?**

* Los vectores se usan para comparar y medir similitud entre objetos.

Por ejemplo, si tenemos dos usuarios:



```python
u1 = np.array([0.8, 0.3, 0.6])
u2 = np.array([0.7, 0.4, 0.5])
```

Podemos medir qué tan parecidos son calculando la similitud del coseno:

```python
from numpy import dot
from numpy.linalg import norm

cosine_similarity = dot(u1, u2) / (norm(u1) * norm(u2))
print("Similitud del coseno:", cosine_similarity)
```

**Interpretación:**

* Si el resultado ≈ 1 → son muy parecidos.

* Si ≈ 0 → no tienen relación.

* Si < 0 → son opuestos.

Este cálculo se usa en sistemas de recomendación (Netflix, Spotify, Amazon), donde los vectores representan a usuarios y productos.

#### **Matrices**

Una **matriz** es una colección de vectores organizados en filas y columnas.

Ejemplo: A = [[1, 2], [3, 4]]

#### **Ejemplo en Python - Operaciones con matrices**

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print("Suma:
", A + B)
print("Multiplicación (producto matricial):
", np.dot(A, B))
```

##**UNIDAD 5: Derivadas y Aprendizaje Automático**

#### **Qué es una derivada**

Una **derivada** mide cómo cambia una función cuando cambia su variable.

Ejemplo: f(x)=x² ⇒ f'(x)=2x


#### **Derivadas en IA: cómo aprenden los modelos?**

Las derivadas son la base del aprendizaje en los modelos de IA.
En especial, se usan para ajustar los parámetros (pesos) de las redes neuronales mediante un proceso llamado retropropagación (backpropagation).

El objetivo de una red neuronal es minimizar el error entre su predicción y la realidad.

Ese error se mide con una función de pérdida (loss).

**Para reducir ese error:**

* Se calcula la derivada del error respecto a cada peso.

* Esa derivada indica en qué dirección debe moverse el peso para mejorar.

* Se actualizan los pesos con un pequeño paso en sentido opuesto a la derivada.




#### **Ejemplo en Python - Cálculo de pérdida y derivada (numérico)**

```python
x, t = 2, 10
w = 3
y = w * x
L = (y - t)**2
derivada = 2 * x * (y - t)
print("Pérdida:", L, " Derivada:", derivada)
```

* Si la derivada es positiva → el modelo debe disminuir w.
* Si es negativa → debe aumentarlo.

Este proceso se repite miles de veces con muchos datos, hasta que los errores se minimizan.

Así es como una red neuronal “aprende”.
