###**Implementacion de una Pila**

En Python, una pila se puede definir de la siguiente manera utilizando el tipo de datos lista.

```python
stack = []

def push(stack, item):
  stack.append(item)

def pop(stack):
  return stack.pop()
```

```python
stack = [] #Creamos nuestra pila vacia
print(stack) #imprimimos en pantalla

push(stack, "A") #Añadimos un elemento a la pila
print(stack)

push(stack, "B") #Añadimos un elemento a la pila
print(stack)


pop(stack) #Eliminamos el elemento de la parte superior
print(stack) #imprimimos la pila

push(stack, "C") #Añadimos un elemento a la pila
print(stack)

pop(stack) #Eliminamos el elemento de la parte superior
print(stack)

pop(stack) #Eliminamos el elemento de la parte superior
print(stack)

pop(stack) #IndexError , ya que no tiene nada que borrar.
print(stack)
```

**IMPORTANTE:** Si llamamos al metodo **.pop()** en una pila vacia nos arrojara un 'IndexError' porque no tiene nada que borrar.

* Parar manejar de mejor manera este tipo de error podemos hacer lo siguiente:

```python
#La funcion is_empty() delvuelve True si la longitud de la pila como parametro es cero y False en caso contrario.

def is_empty(stack):
  return True if len(stack) == 0 else False

stack = []
print(is_empty(stack))
```

```python
def pop(stack):
  return None if is_empty(stack) else stack.pop() #Establecemos una condicion

stack = []
print(stack)
print(pop(stack))
print(stack)

#La funcion pop() borra y devuelve el objeto None si la pila esta vacía, y si la pila no esta vacía, borra y
#devuelve el elemento superior de la lista de la pila.

#La pila inicializada esta vacía, por lo que pop() devuelve el objeto None y no se produce ningun IndexError.
```

##**Tema 02 - Colas**

Una cola es un tipo de datos abstracto en el que sólo es posible la eliminación en la parte delantera y sólo es posible la adición en la parte trasera.

Dado que el primer elemento añadido también se elimina el primero, una cola es una estructura de datos que funciona con el método **FIFO (First-In-First-Out).**

**Una cola debe proporcionar dos interfaces importantes:**

* **enqueue:** añade un nuevo elemento al final de la cola
* **dequeue:** elimina un elemento de la parte delantera de la cola



image.png

**En Python podemos implementarlas con:**

+ **Listas** (list) (pero no son eficientes al sacar del inicio).

+ **deque de collections** (más recomendado).

+ **queue.Queue** (especialmente en programación concurrente o multihilo).

**Ejemplos de colas:**
Es frecuente encontrar una cola en nuestra vida.
* Para reservar un vuelo en un aeropuerto, la gente tiene que esperar en una cola, y la primera persona que ha entrado en ella recibe primero el servicio.
Además, es habitual encontrar colas en restaurantes y bancos.

---

**Sintaxis Basica**

```python
from collections import deque

# Crear una cola
cola = deque()

# Agregar elementos (enqueue)
cola.append("A")
cola.append("B")
cola.append("C")

print("Cola actual:", cola)  # deque(['A', 'B', 'C'])

# Sacar elementos (dequeue)
elemento = cola.popleft() # pop elimina el elemento mas a la derecha pero popleft elimina el elemento mas a la izquierda.
print("Elemento atendido:", elemento)  # A
print("Cola después del dequeue:", cola)  # deque(['B', 'C'])
```

**NOTA:** Se usa append() para encolar y popleft() para desencolar (deque es lo más eficiente).

---

###**Ejemplos:**

**Descargas en espera**

* Cuando descargas varios archivos, se encolan en orden.

* El primero que empezó se atiende primero.

```python
from collections import deque

descargas = deque()
descargas.append("archivo1.zip")
descargas.append("archivo2.mp4")
descargas.append("archivo3.pdf")

print("Cola de descargas:", descargas)

# Procesar la primera descarga
descargado = descargas.popleft()
print("Descargando:", descargado)
print("Cola restante:", descargas)
```

**Atención al cliente (call center o tickets)**

* Las solicitudes se atienden en el orden que llegan.

* No puedes atender al último primero.

```python
clientes = deque()
clientes.append("Cliente 1")
clientes.append("Cliente 2")
clientes.append("Cliente 3")

print("Cola de clientes:", clientes)

# Atender al primero
atendido = clientes.popleft() # es un método de la clase deque que elimina y devuelve el primer elemento (el que está más a la izquierda) de la cola.
print("Atendiendo a:", atendido)
print("Cola restante:", clientes)
```

**Procesamiento de tareas (sistemas operativos)**

* El sistema operativo coloca procesos en una cola para ser ejecutados.

* El CPU atiende en orden, salvo prioridades especiales.

```python
procesos = deque(["Proceso 1", "Proceso 2", "Proceso 3"])

while procesos:
    actual = procesos.popleft()
    print("Ejecutando:", actual)
```

##**Tema 03 -Busqueda Secuencial**

La búsqueda secuencial es el algoritmo más simple para localizar un elemento en una lista, este consiste en recorrer uno por uno los elementos hasta encontrar el valor buscado o llegar al final.

Es directa, fácil de implementar y no requiere que la lista esté ordenada.
Esto la diferencia de la búsqueda binaria, que necesita datos previamente ordenados.

.

**Es útil cuando:**

* Los datos NO están ordenados (a diferencia de la binaria).

* La lista es pequeña o se busca una única vez.

* Se trabaja con estructuras en las que no se puede aplicar acceso directo ordenado (por ejemplo, leer registros de un archivo uno a uno).

**Sintaxis**

```python
def busqueda_secuencial(lista, x):
    for i in range(len(lista)):      # recorrer toda la lista
        if lista[i] == x:            # comparar elemento
            return i                  # si lo encuentra → devuelve posición
    return -1                        # si no está → devuelve -1
```

###**Cuándo usarla (utilidad y limitaciones)**

**Ventajas / Cuándo usarla**

1. Cuando la lista no está ordenada.

2. Si la lista es pequeña (el costo de recorrerla no es grande).

3. Para búsquedas únicas o poco frecuentes.

4. En estructuras donde no se puede acceder aleatoriamente (ejemplo: leer registros de un archivo uno por uno).

**Desventajas / Cuándo evitarla**

1. En listas grandes, porque es lenta comparada con otros algoritmos (O(n)).

2. Si vas a hacer muchas búsquedas sobre la misma lista, conviene ordenarla primero y usar búsqueda binaria (O(log n)).

###**Logica**

1. Recorre la lista desde el índice 0 hasta el último.

2. En cada paso compara el valor de la posición actual con el objetivo x.

3. Si lo encuentra, retorna el índice.

4. Si llega al final sin hallarlo, devuelve -1.

```python
def seq_search(nums, x):
    for i in range(len(nums)):
        print(f"índice={i}, valor={nums[i]}")
        if nums[i] == x:
            return i
    return -1

S = [11, 17, 26, 28, 37, 45, 53, 59]
pos = seq_search(S, 75)
print("Resultado:", pos)
```

###**Relación con algoritmo para encontrar la mayor clave**

En realidad, buscar el valor mayor en una lista también se hace de forma secuencial, revisando elemento por elemento y comparándolo con el máximo actual.

Ejemplo con la lista:
S = [11, 37, 45, 26, 59, 28, 17, 53]

* Inicialmente, largest = 0 → S[0] = 11

* Comparamos:

  * S[1] = 37 > S[0] = 11 → ahora el mayor está en posición 1.

  * S[2] = 45 > 37 → ahora el mayor está en posición 2.

  * … y así hasta terminar.

Al final, el mayor será 59 en la posición 4.

```python
def mayor_clave(lista):
    largest = 0
    for i in range(1, len(lista)):
        if lista[i] > lista[largest]:
            largest = i
    return largest, lista[largest]

S = [11, 37, 45, 26, 59, 28, 17, 53]
pos, valor = mayor_clave(S)
print(f"El mayor es {valor} en la posición {pos}")
```

El algoritmo para encontrar la mayor clave se utiliza cuando queremos identificar el número más grande dentro de una lista de valores. Su lógica es muy sencilla: se recorre la lista elemento por elemento y se compara cada valor con el mayor encontrado hasta el momento; si aparece uno más grande, se actualiza como el nuevo máximo. Esto es útil en muchas situaciones reales, como saber qué estudiante obtuvo la calificación más alta, cuál fue el día con mayores ventas o qué sensor registró la temperatura más elevada.

En pocas palabras, este algoritmo responde a la pregunta: “¿Cuál es el valor máximo de este conjunto de datos?”.

##**Tema 04 - Busqueda Binaria**

La búsqueda binaria es un algoritmo eficiente que se utiliza para encontrar un elemento dentro de una lista ordenada.
La idea principal es dividir y conquistar:

Su finalidad es reducir drásticamente el número de comparaciones: en lugar de revisar uno por uno (búsqueda lineal, O(n)), divide el rango de búsqueda a la mitad en cada paso (O(log n)).

En lugar de revisar todos los elementos uno por uno (como en la búsqueda lineal), la lista se divide en mitades sucesivas hasta encontrar el elemento deseado o determinar que no existe.

**Se usa cuando:**

* Los datos están ordenados (o se pueden ordenar una vez y reutilizar).

* Necesitas consultas rápidas: encontrar usuarios por ID, ubicar un registro en un índice, validar si existe un valor, etc.

image.png

image.png

**¿Por qué existe y para qué se usa?**

>**Problema:** En listas muy grandes, la búsqueda secuencial (lineal) es lenta, pues revisa cada elemento uno por uno → O(n) tiempo.

> **Solución:** La búsqueda binaria reduce drásticamente el número de comparaciones → O(log n) tiempo.

**Usos reales:**

* Bases de datos (buscar registros ordenados).

* Índices en motores de búsqueda.

* Validaciones en grandes volúmenes de datos (como logins en sistemas).

**Sintaxis Basica**

```python
def bin_search(nums, x):
    low, high = 0, len(nums) - 1
    while low <= high:
        mid = (low + high) // 2
        if nums[mid] == x:
            return mid
        elif nums[mid] > x:
            high = mid - 1
        else:
            low = mid + 1
    return -1 # indica que el valor no se encuentra en la lista/arreglo/tupla,etc
```

**La idea central (la lógica)**

+ Mantén un rango de búsqueda limitado por dos índices: low (inicio) y high (fin).

+  Calcula el punto medio mid = (low + high) // 2.

+ Compara el objetivo x con S[mid]:

  + Si S[mid] == x: ¡encontrado!

  + Si S[mid] > x: el objetivo, si existe, está a la izquierda → mueve high = mid - 1.

  + Si S[mid] < x: el objetivo está a la derecha → mueve low = mid + 1.

+ Repite mientras low <= high. Si el bucle termina sin encontrarlo, no existe (-1).

**Invariante mental útil**

> “Si x está en la lista, debe estar dentro del rango [low, high] actual.”
Cada iteración mantiene ese invariante y reduce el rango a la mitad.

**Nota:** En Python no hay problema de overflow, pero una forma segura de medio es mid = low + (high - low) // 2.

**Ejemplo sencillo**

```python
S = [11, 17, 26, 28, 37, 45, 53, 59]
x = 53
pos = bin_search(S, x)

print(f"En S, {x} está en la posición {pos}.")
```

**Ejemplo real**

Supongamos que trabajas en un sistema que guarda los IDs de clientes ordenados y quieres verificar si un cliente existe en la base de datos:

```python
clientes = [1001, 1020, 1035, 1044, 1070, 1100, 1200, 1350, 1400]
id_cliente = 2100
pos = bin_search(clientes, id_cliente)

if pos != -1:
    print(f"Cliente con ID {id_cliente} encontrado en la posición {pos}.")
else:
    print(f"Cliente con ID {id_cliente} no está registrado.")
```

###**Búsqueda Binaria con Análisis de Complejidad**

```python
def bin_search(nums, x):
    low, high = 0, len(nums) - 1
    while low <= high:
        mid = (low + high) // 2
        print(f"Buscando en rango [{low}, {high}], mid = {mid}")
        if nums[mid] == x:
            return mid
        elif nums[mid] > x:
            high = mid - 1
        else:
            low = mid + 1
    return -1

S = [11, 17, 26, 28, 37, 45, 53, 59]
x = 53
print("Posición encontrada:", bin_search(S, x))

print('-'*30)

def seq_search(nums, x):
    for i in range(len(nums)):
        print(f"índice={i}, valor={nums[i]}")
        if nums[i] == x:
            return i
    return -1

S = [11, 17, 26, 28, 37, 45, 53, 59]
pos = seq_search(S, 53)
print("Posición encontrada:", pos)
```

Esto muestra cómo el rango se va partiendo en mitades → confirmando la complejidad logarítmica O(log n).

###**Implementacion con Clases en Python**

Podemos modelar la búsqueda binaria como una clase, lo que es útil en programación orientada a objetos (POO).

```python
class BinarySearch:
    def __init__(self, data):
        self.data = sorted(data)  # Siempre mantener la lista ordenada

    def search(self, x):
        low, high = 0, len(self.data) - 1
        while low <= high:
            mid = (low + high) // 2
            print(f"Buscando en rango [{low}, {high}], mid = {mid}")
            if self.data[mid] == x:
                return mid
            elif self.data[mid] > x:
                high = mid - 1
            else:
                low = mid + 1
        return -1

# Ejemplo de uso
nums = [11, 17, 26, 28, 37, 45, 53, 59]
busqueda = BinarySearch(nums)

print("Resultado:", busqueda.search(53))  # Lo encuentra
print("Resultado:", busqueda.search(100)) # No está
```

La búsqueda binaria es fundamental porque optimiza búsquedas en grandes conjuntos de datos ordenados y su rapidez la hace indispensable en sistemas reales como bases de datos, motores de búsqueda y validaciones masivas.

##**Tema 05 - Tabla Hash**

Una tabla hash (o mapa hash) es una estructura de datos abstracta que permite almacenar y acceder a datos rápidamente, usando un mecanismo llamado hashing.

* Una clave (key) se transforma mediante una función hash en un valor numérico.

* Ese valor indica la posición (bucket o cubo) donde guardar o buscar el dato.

**Las operaciones principales son:**

* put(clave, valor): insertar un valor.

* get(clave): recuperar un valor.

* hash(clave): calcular el índice donde se guarda el valor.

image.png

###**Ejemplo de tabla hash con libros**

Supongamos que queremos almacenar libros.
Para obtener la clave de cada libro, sumamos los códigos ASCII de sus letras.


```python
books = [
    "The Little Prince",
    "The Old Man and the Sea",
    "The Little Mermaid",
    "Beauty and the Beast",
    "The Last Leaf"
]

#Funcion para calcular la clave (KEY)
for book in books:
    key = sum(map(ord, book))  # suma de ASCII
    print(key, book)
```

Aquí cada título tiene un número único (su clave hash base).

###**Hashing y funciones hash**

La función hash convierte una clave grande en un número dentro de un rango fijo (el tamaño de la tabla).

###**Definir una tabla hash como clase**

```python
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = {}
        for i in range(size):
            self.table[i] = []

    def hash(self, key):
        return key % self.size #esto nos regresa la posicion del bucket

    def get(self, key):
        return self.table[self.hash(key)] #Busca en la tabla el valor asociado (hash) a la clave enviada.

    def put(self, key, value):
        bucket = self.table[self.hash(key)]
        if value not in bucket:
            bucket.append(value)
```

* __init __: crea la tabla vacía con 'n' cantidad de buckets.

* hash: calcula la posición.

* put: guarda el valor en su bucket

* get: recupera el valor del bucket correspondiente.

**Ejemplo**

```python
#Crear la tabla
table = HashTable(8)
for key in table.table.keys():
    print(key, table.table[key])
```

```python
# Insertar un libro
book = "The Little Prince" #valor
key = sum(map(ord, book)) #clave
table.put(key, book)

for key in table.table.keys():
    print(key, table.table[key])
```

```python
# Insertar varios libros
books = [
    "The Old Man and the Sea",
    "The Little Mermaid",
    "Beauty and the Beast",
    "The Last Leaf"
]

for book in books:
    key = sum(map(ord, book))
    table.put(key, book)

for key in table.table.keys():
    print(key, table.table[key])
```

```python
# Buscar un libro
title = "The Old Man and the Sea"
key = sum(map(ord, title))
bucket = table.get(key)
print(key, bucket)
```

**¿Por qué son útiles las tablas hash?**

* Acceso a datos en tiempo casi constante O(1).

* Muy usadas en diccionarios, índices de bases de datos, caches.

* Soportan grandes volúmenes de información de forma eficiente.

###**Resumen:**

* Definición: Estructura clave–valor con hashing.

* Ejemplo: Libros → suma de ASCII como clave.

* Hashing: f(x) = x % tamaño.

* Clase HashTable: métodos put, get, hash.

* Aplicación: Inserción, colisiones y búsqueda de valores.
