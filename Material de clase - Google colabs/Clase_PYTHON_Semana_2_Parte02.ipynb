## **TEMA 01 - Uso de la herramienta Turtle Graphics**

El m√≥dulo turtle de Python permite dibujar de forma gr√°fica moviendo una "tortuga" sobre una ventana, como si fuera un l√°piz que traza l√≠neas.
Se usa mucho para aprender programaci√≥n visualmente.

```python
import turtle

# Crear la ventana y la tortuga
t = turtle.Turtle()

# Dibujar un cuadrado
for _ in range(4):
    t.forward(100)  # avanza 100 pasos
    t.right(90)     # gira 90 grados

turtle.done()
```

La tortuga avanza (forward), retrocede (backward), gira (right, left), cambia color, grosor, etc.

##**TEMA 02 - Expresiones regulares y metacaracteres en Python (re)**

Una expresi√≥n regular (abreviada como regex) es una cadena de texto especial que define un patr√≥n de b√∫squeda.

> Sirve para buscar, validar, extraer o reemplazar fragmentos de texto que cumplen ciertas reglas.

* En lugar de buscar texto exacto, las expresiones regulares permiten decir:

* "Quiero cualquier n√∫mero de 3 d√≠gitos".

* "Quiero una palabra que empiece con may√∫scula".

* "Quiero todas las direcciones de correo dentro de este texto".

**Ejemplos cotidianos**

**Validar un n√∫mero de tel√©fono**

* Expresi√≥n regular: \d{3}-\d{4}

* **Significado:** tres n√∫meros, guion, cuatro n√∫meros.

* Texto: "Llama al 555-1234" ‚úÖ coincide.
---
**Encontrar correos electr√≥nicos**

* Expresi√≥n regular: \w+@\w+\.\w+

* Texto: "Escribe a prueba@gmail.com" ‚úÖ coincide.
---
**Buscar todas las palabras que empiezan con may√∫scula**

* Expresi√≥n regular: \b[A-Z]\w*

* Texto: "Hola mundo, Soy estudiante del SIC" ‚Üí ‚úÖ encuentra "Hola", "Soy", "SIC".

---

### **Modulo re**

El m√≥dulo re permite buscar, validar y extraer texto usando expresiones regulares (regex).

**Funciones importantes:**

* re.search(patron, texto) ‚Üí Busca la primera coincidencia.

* re.findall(patron, texto) ‚Üí Devuelve todas las coincidencias en una lista.

* re.match(patron, texto) ‚Üí Solo busca al inicio del texto.

* re.split(patron, texto) ‚Üí Divide un texto en partes.

* re.sub(patron, reemplazo, texto) ‚Üí Reemplaza coincidencias.
---
**M√©todos de los objetos encontrados:**

* .group() ‚Üí Devuelve el texto encontrado.

* .start() ‚Üí √çndice donde inicia la coincidencia.

* .end() ‚Üí √çndice donde termina.

* .span() ‚Üí Tupla (inicio, fin).

```python
import re

texto = "Mis correos son test1@gmail.com y ejemplo123@yahoo.com"

# Buscar todos los correos
patron = r"\w+@\w+\.\w+"
coincidencias = re.findall(patron, texto)

print(coincidencias)
```

```python
import re

texto = "Tel√©fonos: 123-456, 987-654, 555-111"
patron = r"\d{3}-\d{3}"

print(re.findall(patron, texto))
```

```python
import re

texto = "Hola123, chau456"

# Buscar todos los n√∫meros
print(re.findall(r"\d+", texto))  #un metacaracter que significa "uno o m√°s d√≠gitos seguidos".
# üëâ ['123', '456']

# Buscar todas las palabras
print(re.findall(r"\w+", texto))  #un metacaracter que significa "una palabra compuesta de letras/n√∫meros".
# üëâ ['Hola123', 'chau456']
```

```python
import re

text = 'please call 010-2345-1234'
regex = re.compile('(\d{3})-(\d{4}-\d{4})')
match_obj = regex.search(text)

print(match_obj.group())   # 010-2345-1234
print(match_obj.group(1))  # 010
print(match_obj.group(2))  # 2345-1234
```

¬øQu√© pasa aqu√≠?

\d{3} ‚Üí encuentra 3 d√≠gitos (010)

\d{4} ‚Üí encuentra 4 d√≠gitos (2345 y 1234)

Los par√©ntesis () crean grupos para separar coincidencias.

.group(0) o .group() ‚Üí el match completo ‚Üí "010-2345-1234"

.group(1) ‚Üí primer grupo ‚Üí "010"

.group(2) ‚Üí segundo grupo ‚Üí "2345-1234"

###**Que es un metacaracter?**

Un metacaracter es un s√≠mbolo especial dentro de las expresiones regulares que no representa literalmente un car√°cter, sino que tiene un significado especial para definir un patr√≥n de b√∫squeda.

Dicho de otra forma:

* Si escribes una letra normal (ejemplo: a), la expresi√≥n regular buscar√° la letra a exactamente.

* Pero si usas un metacaracter (ejemplo: \d), no busca la letra d, sino que significa ‚Äúcualquier n√∫mero del 0 al 9‚Äù.

**Metacaracteres mas comunes:**

* ^ ‚Üí Inicio de cadena
* $ ‚Üí Fin de cadena
* . ‚Üí Cualquier car√°cter
* \d ‚Üí N√∫mero (0-9)
* \w ‚Üí Caracter o n√∫mero
* \s ‚Üí Espacio, tab, salto de l√≠nea
* ' * ‚Üí 0 o m√°s repeticiones
* ' + ‚Üí 1 o m√°s repeticiones
* [abc] ‚Üí a, b o c
* [^abc] ‚Üí Cualquier cosa menos a, b, c

##**Programaci√≥n procedimental**

> La programaci√≥n procedimental es un paradigma que organiza el c√≥digo en funciones o procedimientos que se ejecutan de forma secuencial.
Los datos se procesan a trav√©s de esas funciones, pero los datos y las funciones est√°n separados.

Esto funciona muy bien en programas peque√±os o medianos, pero cuando el proyecto crece y hay muchos datos y funciones, la gesti√≥n se complica. Modificar una parte puede afectar otras, lo que dificulta el mantenimiento.

**Ejemplo mental:**
* Piensa en una receta de cocina. Cada paso (funci√≥n) se ejecuta uno tras otro, y los ingredientes (datos) se pasan a cada paso.

##**Programaci√≥n orientada a objetos (POO)**

> La programaci√≥n orientada a objetos surge para solucionar esas limitaciones.
En este paradigma, el c√≥digo se organiza en objetos, que son unidades que combinan datos y comportamientos en un mismo lugar.

En lugar de tener funciones separadas que trabajan sobre datos sueltos, ahora tenemos objetos que ‚Äúsaben‚Äù cu√°les son sus datos y c√≥mo trabajar con ellos.

Esto permite una mejor organizaci√≥n, mayor reutilizaci√≥n del c√≥digo y facilita mantener o ampliar programas grandes.

**Ejemplo mental:**
* Si en vez de pensar en una receta de cocina piensas en un "robot chef", el robot ya sabe cu√°les son sus ingredientes (datos) y qu√© acciones puede hacer (m√©todos). No hay que pasarle todo desde fuera: el objeto ‚Äúcoche‚Äù, ‚Äúpersona‚Äù o ‚Äúcuenta bancaria‚Äù ya tiene dentro lo necesario.

#**TEMA 04 - Creacion y uso de Clases en Python**

Una clase en Python es un molde o plantilla que permite crear objetos.
En la clase se definen las caracter√≠sticas (atributos) y los comportamientos (m√©todos) que tendr√°n los objetos creados a partir de ella.

> Gracias a las clases podemos modelar entidades del mundo real (como personas, autos o productos) dentro de un programa, manteniendo el c√≥digo m√°s organizado, reutilizable y f√°cil de mantener.

**¬øPara qu√© se usan las clases?**

* Para agrupar datos y funciones que est√°n relacionadas.

* Para modelar entidades reales o abstractas (ejemplo: estudiantes, cuentas bancarias, jugadores en un videojuego).

* Para crear m√∫ltiples objetos basados en una misma definici√≥n.

* Para organizar programas grandes en piezas m√°s manejables.

**Sintaxis Basica**

```python
class NombreClase: #Definicion de la clase
    def __init__(self, parametros):
        self.atributo1 = valor
        self.atributo2 = valor

    def metodo(self):
        # Acci√≥n que realiza el objeto
        pass
```

**Elementos que componen una clase:**

* Nombre de la clase: por convenci√≥n, inicia con may√∫scula (Persona, Coche).

* Constructor __init__: m√©todo especial que se ejecuta al crear el objeto; inicializa atributos.

* Atributos: variables propias del objeto que guardan informaci√≥n (ejemplo: nombre, edad, color).

* M√©todos: funciones definidas dentro de la clase que describen lo que el objeto puede hacer.

* Objetos: instancias creadas a partir de la clase, cada uno con sus propios valores.

**Ejemplos:**

```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre   # atributo
        self.edad = edad       # atributo

    def saludar(self):         # m√©todo
        print(f"Hola, soy {self.nombre} y tengo {self.edad} a√±os.")
```

```python
p1 = Persona("Ana", 25)
p2 = Persona("Luis", 30)

p1.saludar()  # Hola, soy Ana y tengo 25 a√±os.
p2.saludar()  # Hola, soy Luis y tengo 30 a√±os.
```

**¬øQu√© es self?**

self es una referencia al propio objeto que se est√° creando o usando.

Cuando instancias una clase, cada objeto necesita almacenar sus propios datos.

self permite que cada m√©todo sepa a qu√© objeto pertenece y pueda acceder a sus atributos (variables) y m√©todos (funciones dentro de la clase).

.

**En otras palabras:**

self es el "yo" del objeto. Si creas dos coches, cada uno tiene su propio "yo" (su propia marca, color, velocidad, etc.).


**¬øPor qu√© siempre hay que ponerlo?**

* Python no mete self autom√°ticamente, t√∫ lo defines en los m√©todos.

* La primera posici√≥n en los m√©todos de una clase siempre corresponde a la instancia que invoca ese m√©todo.

* Aunque escribas mi_coche.acelerar(20), internamente Python traduce eso a:


      Coche.acelerar(mi_coche, 20)



Es decir, el objeto mi_coche se pasa como primer argumento al m√©todo, y ese primer argumento lo llamamos self por convenci√≥n.

```python
class Coche:
    def __init__(self, marca, color):
        self.marca = marca     # ‚Üê guardo marca en este objeto
        self.color = color     # ‚Üê guardo color en este objeto
        self.velocidad = 0     # ‚Üê cada objeto empieza con velocidad 0

    def acelerar(self, cantidad):
        self.velocidad += cantidad   # ‚Üê modifico la velocidad de "este coche"
        print(f"El {self.marca} aceler√≥ a {self.velocidad} km/h")

    def frenar(self):
        self.velocidad = 0
        print(f"El {self.marca} se detuvo")
```

```python
mi_auto = Coche("Toyota", "Rojo")

mi_auto.acelerar(50)   #  El Toyota aceler√≥ a 50 km/h
mi_auto.frenar()       #  El Toyota se detuvo

c1 = Coche("Toyota", "Rojo")
c2 = Coche("Mazda", "Azul")

c1.acelerar(30)   # afecta solo a c1
c2.acelerar(50)   # afecta solo a c2
```

* Es necesario para guardar y acceder a los atributos y m√©todos de cada objeto.

* Sin self, tus m√©todos no sabr√≠an a qu√© instancia aplicarse.

##**Constructores en Python**

###**Metodo __init __**

Un constructor es un m√©todo especial dentro de una clase que se encarga de crear e inicializar un objeto.
En Python, ese constructor se llama __init __.

**Cuando escribimos:**

     objeto = Clase(parametros)

Python autom√°ticamente llama al m√©todo __init__ de esa clase, pasando los valores que le damos como par√°metros.

**¬øPara qu√© sirven los constructores?**

* **Inicializar los atributos del objeto:** Al momento de crear un objeto, el constructor permite asignarle valores iniciales (nombre, edad, color, etc.).

* **Ahorrar c√≥digo repetitivo:** En lugar de tener que asignar atributos uno por uno despu√©s de crear el objeto, el constructor lo hace todo en un solo paso.

* **Garantizar consistencia:** Gracias al constructor, sabemos que todo objeto creado tendr√° los atributos necesarios, evitando errores.

* **Dar valores por defecto:** Se pueden establecer valores predeterminados para los atributos, en caso de que no se proporcionen.

```python
"""
Cuando escribimos Persona("Ana", 25), Python crea un objeto de tipo Persona.
Luego ejecuta el constructor __init__ y guarda "Ana" y 25 dentro de los atributos nombre y edad.

"""

class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre   # atributo inicializado
        self.edad = edad       # atributo inicializado

# Crear objetos usando el constructor
p1 = Persona("Ana", 25)
p2 = Persona("Luis", 30)

print(p1.nombre, p1.edad)  # Ana 25
print(p2.nombre, p2.edad)  # Luis 30

```

**¬øPor qu√© se usan los constructores?**

* Porque permiten crear objetos completos desde el inicio.

* Porque hacen que el c√≥digo sea m√°s ordenado y claro.

* Porque son una de las bases de la Programaci√≥n Orientada a Objetos (POO) ya que ayudan a que cada objeto tenga su propia identidad desde el momento en que nace.

```python
class Gato:
    def __init__(self, nombre="Michi", color="Gris"):
        self.nombre = nombre
        self.color = color

g1 = Gato()                     # Usa los valores por defecto
g2 = Gato("Pelusa", "Blanco")   # Usa valores personalizados

print(g1.nombre, g1.color)  # Michi Gris
print(g2.nombre, g2.color)  # Pelusa Blanco
```

###**Metodo __str __**

El m√©todo __str__ es un m√©todo especial en Python que define c√≥mo se debe mostrar un objeto en forma de texto.

En otras palabras, permite que cuando hagamos print(objeto) veamos informaci√≥n clara y legible, en lugar de un mensaje t√©cnico como:

```python
<__main__.Clase object at 0x000001F4>
```

**¬øPara qu√© sirve __str __?**

* Hacer m√°s legible la informaci√≥n de un objeto: En vez de ver direcciones de memoria, obtenemos una descripci√≥n con sentido.

* Facilitar la depuraci√≥n (debugging) : Si un programa imprime objetos en la consola, __str __ nos ayuda a saber qu√© contiene cada objeto.

* Personalizar la representaci√≥n textual: T√∫ decides qu√© atributos mostrar y c√≥mo se ver√°n.

**¬øPor qu√© se usa __str __?**

* Porque sin √©l, Python no sabe c√≥mo mostrar el objeto de manera entendible.

* Porque permite representar al objeto en forma de cadena (texto) de la manera que nosotros necesitemos.

* Porque mejora la experiencia de lectura cuando trabajamos con muchos objetos.

**SINTAXIS**

```python
def __str__(self):
    return "Texto que represente al objeto"

"""
Siempre debe devolver un string (str).
Dentro del string podemos incluir atributos del objeto.
"""
```

**Ejemplo**

```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def __str__(self):
        return f"Persona: {self.nombre}, Edad: {self.edad}"
```

```python
p1 = Persona("Ana", 25)
p2 = Persona("Luis", 30)

print(p1)  # Persona: Ana, Edad: 25
print(p2)  # Persona: Luis, Edad: 30

# Gracias a __str__, cuando imprimimos p1 o p2, Python no muestra la direcci√≥n de memoria, sino un texto claro.
```

#**TEMA 05 - Variables de Instancia, Clase, Miembro Y Campo**

Cuando trabajamos con Programaci√≥n Orientada a Objetos (POO), los atributos (variables) pueden ser de distintos tipos.

##**Variables de instancia**

* Son las m√°s comunes.
* Pertenecen a cada objeto en particular.
* Se definen dentro del constructor __init __ con self.
* Cada instancia tiene sus propios valores independientes.

```python
class Gato:
    def __init__(self, nombre, color):
        self.nombre = nombre   # variable de instancia
        self.color = color     # variable de instancia

nabi = Gato("Nabi", "Negro")
nero = Gato("Nero", "Blanco")

print(nabi.nombre)  # Nabi
print(nero.nombre)  # Nero
```

image.png

##**Variables de clase**

* Son compartidas por todas las instancias de la clase.
* Se definen fuera del __init __, directamente dentro de la clase.
* Cambiar su valor desde la clase afecta a todos los objetos.

```python
class Gato:
    especie = "Felino"   # variable de clase

    def __init__(self, nombre, color):
        self.nombre = nombre
        self.color = color

nabi = Gato("Nabi", "Negro")
nero = Gato("Nero", "Blanco")

print(nabi.especie)  # Felino
print(nero.especie)  # Felino

# Aqu√≠, especie es la misma para todos los gatos, porque es una variable de clase.
```

##**Variables Miembro**

El t√©rmino variable miembro se refiere de manera general a cualquier variable asociada a una clase u objeto.

Por eso puede englobar tanto a las variables de instancia como a las variables de clase.

**En pocas palabras:**

> Variables miembro = atributos de la clase u objetos.

##**Campos**

El t√©rmino campo es otro sin√≥nimo muy usado para hablar de atributos en POO.

Al igual que ‚Äúvariable miembro‚Äù, hace referencia a los datos que pertenecen a una clase u objeto.

Ejemplo: en la clase Gato, los campos son nombre y color.

# **TEMA 06 - Encapsulacion en clases**

La encapsulaci√≥n es un principio de la programaci√≥n orientada a objetos (POO) que consiste en proteger los atributos y m√©todos de una clase para que no sean accedidos o modificados directamente desde fuera del objeto.

Esto se hace para garantizar que los datos siempre tengan valores v√°lidos y que la interacci√≥n con ellos se haga de forma controlada.

**En Python se utiliza principalmente:**

Convenci√≥n de nombres:

* _atributo (protegido, acceso interno preferido).

* __atributo (privado, fuertemente restringido).

Estos son m√©todos especiales llamados **getters y setters** para leer y modificar estos atributos de forma segura.

* Un getter es un m√©todo que devuelve el valor de un atributo.

* Un setter es un m√©todo que asigna un valor a un atributo, pero puede incluir validaciones.

* Evita modificar atributos de manera directa que podr√≠a romper la l√≥gica interna.

* Permite controlar qu√© valores son aceptados o no.

**Sintaxis / Estructura**

# **TEMA 07 - Herencia en clases**

La herencia permite que una clase (llamada subclase o clase hija) herede atributos y m√©todos de otra clase (llamada superclase o clase padre).

Es uno de los pilares de la POO y sirve para reutilizar c√≥digo y modelar relaciones jer√°rquicas.


---
.
* Permite crear nuevas clases a partir de clases existentes.

* La clase hija hereda atributos y m√©todos de la clase padre.

* Se pueden sobrescribir (override) m√©todos para cambiar su comportamiento.

* Se logra usando par√©ntesis con el nombre de la clase padre.

image.png

**Sintaxis / Estructura**

```python
class ClasePadre:
    def metodo_padre(self):
        pass

class ClaseHija(ClasePadre):
    def metodo_hija(self):
        pass
```

**Ejemplos**

```python
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def hablar(self):
        print("Hace un sonido")

class Perro(Animal):
    def hablar(self):   # sobrescritura
        print("Guau!")

p = Perro("Firulais")
p.hablar()  # Guau!
```

```python
class Vehiculo:
    def __init__(self, marca):
        self.marca = marca

class Coche(Vehiculo):
    def __init__(self, marca, color):
        super().__init__(marca)   # herencia del constructor
        self.color = color

auto = Coche("Toyota", "Rojo")
print(auto.marca, auto.color)  # Toyota Rojo
```

# **TEMA 08 - Jerarquia en clases**

Una jerarqu√≠a de clases es la organizaci√≥n de clases en niveles, donde las clases m√°s generales est√°n arriba y las m√°s espec√≠ficas abajo.

Se construye mediante herencia m√∫ltiple o simple, creando un √°rbol de relaciones.

* Permite representar relaciones de tipo ‚Äúes un‚Äù (ejemplo: un perro es un animal).

* Se organiza desde lo m√°s gen√©rico (superclase) hasta lo m√°s espec√≠fico (subclases).

* Favorece el reuso y extensi√≥n del c√≥digo.

* Es com√∫n en sistemas grandes con m√∫ltiples entidades.

**Sintaxis / Estructura**

```python
class SuperClase:
    pass

class SubClase(SuperClase):
    pass

class SubSubClase(SubClase):
    pass
```

**Ejemplos**

```python
class Animal:
    pass

class Mamifero(Animal):
    pass

class Perro(Mamifero):
    pass

mi_perro = Perro()
print(isinstance(mi_perro, Animal))   # True
```
