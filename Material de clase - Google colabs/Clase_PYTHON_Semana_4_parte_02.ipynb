##**ENFOQUE GREEDY**

El enfoque Greedy (codicioso, voraz) es una **tecnica de diseno de algoritmos** que llega a una solucion seleccionando la "mejor" eleccion del momento de una secuencia de elecciones segun criterios predefinidos.

> Es una técnica de diseño de algoritmos muy eficiente y sencilla para resolver problemas de optimización, como el problema del cambio de moneda y el de la selección de actividades.

Algunos algoritmos voraces populares son Fractional Knapsack , el algoritmo de Dijkstra , el algoritmo de Kruskal , la codificación de Huffman y el algoritmo de Prim.

image.png

En cada etapa este elige la opción más beneficiosa en el momento, sin considerar las implicaciones futuras de esa elección, buscando siempre la mejor alternativa inmediata.

**Características principales:**

**Decisiones locales:**
* Se enfoca en la mejor opción en el presente, sin mirar el panorama completo o las consecuencias a largo plazo.

**Optimización de problemas:**

* Es útil en problemas de optimización donde se busca maximizar o minimizar algo.

**Simplicidad y eficiencia:**

* Suele ser fácil de diseñar y verificar su funcionamiento, aunque no siempre garantiza una solución óptima global.

**¿Cuándo usarlo?**

Un algoritmo greedy puede ser apropiado si se cumplen dos condiciones:

* **Propiedad Greedy:** Elegir la opción óptima en cada paso, por sí sola, lleva a la solución global óptima.
* **Subestructura Óptima:** El problema puede dividirse en subproblemas más pequeños, cuya solución óptima contribuye a la solución general.

**Ejemplo:**

Un ejemplo clásico es el problema del cambio de moneda. Si un algoritmo greedy necesita hacer 30 con monedas de 5, 10 y 25, siempre seleccionará la moneda de mayor valor que no exceda el monto restante. En el caso de un sistema monetario con esas denominaciones, este enfoque sería óptimo.

Sin embargo, no siempre garantiza una solución óptima para cualquier conjunto de monedas.

###**Que es un algoritmo codicioso**

Un algoritmo codicioso se refiere a cualquier algoritmo empleado para resolver un problema de optimización donde el algoritmo procede haciendo una elección localmente óptima (es decir, una elección codiciosa) con la esperanza de que resulte en una solución globalmente óptima.

**Características del algoritmo voraz**

* Los algoritmos codiciosos toman la mejor decisión en cada paso sin reconsiderar.
* Las decisiones tomadas son definitivas y no se pueden cambiar.
* Es posible que no siempre encuentren la mejor solución y en ocasiones puedan obtener resultados subóptimos.
* A pesar de esto, son rápidos en comparación con otros métodos como la programación dinámica.
* Los algoritmos codiciosos son intuitivos, pero es difícil demostrar su corrección matemáticamente.
* En algunos casos, pueden proporcionar soluciones casi óptimas cuando una solución exacta no es práctica.

###**Propiedades del algoritmo**

**Propiedad de elección codiciosa**

Un problema tiene esta propiedad cuando la mejor solución global se puede lograr tomando siempre la mejor opción en cada paso (la elección local más conveniente).

Es decir: en cada paso eliges lo mejor sin preocuparte por lo que vendrá después.

Ejemplo: para dar cambio de 76 con monedas {50, 20, 5, 1}, eliges primero la moneda más grande que no se pase:

1. 76 → tomas 50 → queda 26

2. 26 → tomas 20 → queda 6

3. 6 → tomas 5 → queda 1

4. 1 → tomas 1 → queda 0

**Solución = {50, 20, 5, 1}, y resulta que es la mínima cantidad de monedas.**

**Propiedad de subestructura óptima**

Un problema tiene esta propiedad cuando la solución óptima total se puede construir a partir de las soluciones óptimas de los subproblemas.

**Siguiendo el ejemplo:**

1. Elección codiciosa: primero tomo 50.

2. Subproblema: resolver el cambio de 26.

3. Para 26, el algoritmo da {20, 5, 1}.

**Combino: {50} ∪ {20, 5, 1} = {50, 20, 5, 1}.**

###**Ejercicio del cambio de monedas**

```python
def coin_change(coins, amount):
    changes = []                    # lista donde guardamos las monedas seleccionadas
    largest = 0                       # índice del tipo de moneda que estamos probando (empezamos por la mayor)
    while amount > 0:
        if largest >= len(coins):
            raise ValueError(f"No se puede completar el cambio restante = {amount} con las monedas dadas")
        if amount < coins[largest]:
            largest += 1             # pasamos a la siguiente moneda (más pequeña)
        else:
            changes.append(coins[largest])  # añadimos la moneda actual a la solución
            amount -= coins[largest]        # restamos su valor del resto por cubrir
    return changes                    # devolvemos la lista de monedas elegidas
```

```python
coins = [1,5,10,20,50]
amount = 76

print(coin_change(coins,amount))
```

###**Ejercicio de seleccion de actividades**

**Planteamiento:** Tenemos 7 solicitudes de reuniones, cada una con hora de inicio y hora de fin

image.png

**Pregunta: ¿cuál es el máximo número de reuniones que se pueden atender sin que se solapen?**

**La estrategia es:**

1. Ordenar las reuniones por su hora de finalización (menor → mayor).

2. Seleccionar la primera que termina más temprano.

3. De ahí en adelante, elegir la siguiente reunión cuya hora de inicio sea ≥ hora de fin de la última seleccionada.

4. Repetir hasta agotar todas las solicitudes.

**Selección Greedy**

* Escogemos la primera (id=0, 1→2).

* Siguiente reunión que empieza después o igual a 2 → (id=1, 3→4).

* Siguiente reunión que empieza después o igual a 4 → (id=4, 5→6).

* Siguiente reunión que empieza después o igual a 6 → (id=5, 8→9).

```python
def activity_selection(start, end):
    # start: lista con horas de inicio
    # end: lista con horas de fin
    n = len(start)
    activities = list(zip(range(n), start, end))  # (id, inicio, fin)

    # Paso 1: ordenar por hora de fin
    activities.sort(key=lambda x: x[2])

    selected = []             # aquí guardaremos las reuniones seleccionadas
    last_end = 0              # lleva el final de la última reunión aceptada

    # Paso 2: aplicar selección codiciosa
    for act in activities:
        act_id, s, f = act
        if s >= last_end:     # si no hay solapamiento
            selected.append(act_id)
            last_end = f      # actualizar la hora de fin
    return selected

# Datos del ejemplo
start = [1, 3, 2, 1, 5, 8, 5]
end   = [2, 4, 5, 6, 6, 9, 9]

resultado = activity_selection(start, end)
print("Reuniones seleccionadas:", resultado)
```

##**DIVIDE Y CONQUISTA**

Es una técnica de diseño de algoritmos donde un problema grande se resuelve dividiéndolo en partes más pequeñas, resolviendo esas partes, y luego uniendo los resultados.

**Pasos principales**

1. Dividir

    * Se toma el problema original y se divide en subproblemas más pequeños.

**Ejemplo:** si quieres ordenar una lista grande, divídela en dos mitades.

2. Conquistar

    * Resuelve cada subproblema.

    * Si los subproblemas aún son grandes, se siguen dividiendo (aquí entra la recursividad).

    * Si son lo suficientemente pequeños, se resuelven directamente (caso base).

3. Combinar

    * Une las soluciones de los subproblemas para obtener la solución del problema original.

**Ejemplo:** al ordenar listas, combinas dos listas ordenadas en una sola lista ordenada.

**Ejemplos clásicos de algoritmos Divide y Vencerás**

**Merge Sort (Ordenamiento por fusión):**

    * Divide la lista en mitades, ordena cada mitad, y luego las combina ordenadas.

**Quick Sort (Ordenamiento rápido):**

    * Divide la lista según un pivote (menores a la izquierda, mayores a la derecha), ordena cada parte recursivamente.

**Búsqueda Binaria:**

    * Divide la lista en dos mitades y busca solo en la mitad donde puede estar el elemento.

    * Multiplicación rápida de matrices (Strassen).

🔹 Ventajas

* Divide y Vencerás suele ser más eficiente que resolver el problema de forma directa.

* Aprovecha la recursividad y permite resolver problemas complejos de forma sistemática.

🔹 Desventajas

* El uso intensivo de recursión puede consumir más memoria en la pila de llamadas.

* A veces dividir y combinar añade costos extras (ej. en Merge Sort la fusión).

##**Algoritmo de Dijkstra (camino más corto) - Grafos**

###**¿Qué es?**

Es un algoritmo para encontrar la distancia mínima desde un nodo origen hasta todos los demás nodos en un grafo con pesos positivos.



###**Cómo funciona**

1. Inicializa todas las distancias en ∞, excepto el nodo origen (0).

2. Escoge el nodo con la distancia más pequeña (Greedy).

3. Actualiza (relaja) las distancias de sus vecinos.

4. Marca el nodo como “visitado”.

5. Repite hasta procesar todos los nodos.

###**Estructura**

Entrada: un grafo con nodos, aristas y pesos.

Salida: lista/diccionario con la distancia mínima desde el origen a cada nodo.

###**Código en Python**

```python
import heapq

def dijkstra(grafo, inicio):
    # Inicializamos distancias: todos en infinito excepto el nodo de inicio
    distancias = {nodo: float("inf") for nodo in grafo}
    distancias[inicio] = 0

    # Usamos una cola de prioridad (Greedy: siempre saca el nodo más cercano)
    cola = [(0, inicio)]

    while cola:
        distancia_actual, nodo_actual = heapq.heappop(cola)

        # Si ya tenemos una mejor distancia, lo ignoramos
        if distancia_actual > distancias[nodo_actual]:
            continue

        # Revisamos vecinos
        for vecino, peso in grafo[nodo_actual]:
            nueva_distancia = distancia_actual + peso
            if nueva_distancia < distancias[vecino]:  # Relajación
                distancias[vecino] = nueva_distancia
                heapq.heappush(cola, (nueva_distancia, vecino))

    return distancias

# Ejemplo de grafo representado con diccionario
grafo = {
    "A": [("B", 2), ("C", 5)],
    "B": [("A", 2), ("D", 3)],
    "C": [("A", 5), ("D", 2)],
    "D": [("B", 3), ("C", 2)]
}

print(dijkstra(grafo, "A"))
```

###**Aplicación Greedy**

En cada iteración, Dijkstra elige codiciosamente el nodo con la menor distancia conocida hasta ahora. Esa elección local asegura el resultado global óptimo.

##**Algoritmo de Kruskal (árbol de expansión mínima - MST)**

###**¿Qué es?**

Es un algoritmo que construye un árbol de expansión mínima (MST): conecta todos los nodos de un grafo con el menor costo posible y sin ciclos.



###**Cómo funciona (pasos)**

1. Ordena todas las aristas por peso (menor → mayor).

2. Inicializa un conjunto para representar cada nodo (usando Union-Find).

3. Itera sobre las aristas:

    * Si conecta dos conjuntos diferentes, añádela al MST.

    * Si forma ciclo, se descarta.

    * Termina cuando todos los nodos están conectados.

###**Estructura**

* Entrada: conjunto de nodos y aristas con pesos.

* Salida: lista de aristas que forman el MST.

###**Codigo**

```python
# Estructura de Union-Find (para manejar conjuntos)
class UnionFind:
    def __init__(self, n):
        self.padre = list(range(n))

    def find(self, x):
        if self.padre[x] != x:
            self.padre[x] = self.find(self.padre[x])
        return self.padre[x]

    def union(self, x, y):
        raizX, raizY = self.find(x), self.find(y)
        if raizX != raizY:
            self.padre[raizY] = raizX
            return True
        return False

def kruskal(nodos, aristas):
    # Ordenamos las aristas por peso (Greedy)
    aristas.sort(key=lambda x: x[2])

    uf = UnionFind(nodos)
    mst = []

    for u, v, peso in aristas:
        if uf.union(u, v):  # Si no hay ciclo, se agrega al MST
            mst.append((u, v, peso))

    return mst

# Ejemplo: grafo con 4 nodos (0,1,2,3)
aristas = [
    (0, 1, 1),
    (1, 2, 3),
    (2, 3, 2),
    (0, 2, 4)
]

print(kruskal(4, aristas))
```

###**Aplicación Greedy**

Kruskal es Greedy porque elige siempre la arista más barata disponible que no forme ciclo. Cada elección local contribuye al MST global óptimo.
