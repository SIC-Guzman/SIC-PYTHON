```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import wb #USo de datareader con banco mundial

df = wb.download(indicator="SP.POP.TOTL", country="GT", start=2015, end=2020)

df.head()
df = df.reset_index()
df.head()

plt.plot(df["year"], df["SP.POP.TOTL"])
plt.title("Población de Guatemala")
plt.xlabel("Años")
plt.ylabel("Población")
plt.grid(True)
plt.show()

print('\n')

print('GRAFICO MODIFICADO')

plt.plot(df["year"], df["SP.POP.TOTL"], color='red', linewidth=5) #definimos color de la linea y grosor
plt.title("Población de Guatemala")
plt.xlabel("Años")
plt.ylabel("Población")
plt.grid(True)
plt.show()
```

```python
#GRAFICO DE BARRAS

plt.bar(df["year"], df["SP.POP.TOTL"])
plt.title("Población de Guatemala")
plt.xlabel("Años")
plt.ylabel("Población")
plt.grid(True)
plt.show()
```
