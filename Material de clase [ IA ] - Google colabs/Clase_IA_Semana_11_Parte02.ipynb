###**DEFINICIONES**

> **El concepto de aprendizaje no supervisado**

Es una técnica de aprendizaje automático que sólo se realiza si se dan una serie de variables como X1, X2, X3, ... Xp sin ninguna variable objetivo (o variable de respuesta Y) en el conjunto de datos.

Al no haber variables objetivas (o variables de respuesta) asociadas a variables  explicativas, el objetivo del aprendizaje no supervisado es descubrir un patron especifico o un conocimiento desconocido a partir de unos datos dados en lugar de hacer una predicción.

.

* **Asociacion:** se refiere a descubrir una relacion de co-ocurrencia entre elementos especificos en el conjunto de

###**ANALISIS DE CONGLOMERADOS**

En Inteligencia Artificial (IA) y aprendizaje automático (machine learning), el análisis de conglomerados (también llamado análisis de clusters o clustering) es una técnica de aprendizaje no supervisado que sirve para agrupar datos similares entre sí según sus características.

**El análisis de conglomerados busca dividir un conjunto de datos en grupos (conglomerados o clusters), de modo que:**

* Los elementos dentro del mismo grupo sean muy parecidos entre sí.

* Los elementos de grupos distintos sean lo más diferentes posible.

image.png

###**EJEMPLO:**

Imagina que tienes datos de clientes de una tienda:

* Edad

* Ingreso mensual

* Frecuencia de compra

**El análisis de conglomerados puede descubrir grupos naturales, como:**

* Jóvenes con ingresos bajos que compran con frecuencia.

* Adultos con ingresos altos que compran pocas veces.

* Personas mayores con ingresos medios y compras ocasionales.

**Estos grupos no estaban definidos de antemano: el algoritmo los descubre automáticamente.**

---

###**Cómo funciona:**

1. Seleccionas las variables que describen los datos (por ejemplo, edad e ingreso).

2. El algoritmo mide la similitud o distancia entre los puntos (usando, por ejemplo, la distancia euclidiana).

3. Agrupa los datos de forma que los elementos más cercanos formen un mismo cluster.

4. Evalúas y visualizas los resultados (por ejemplo, con gráficos o métricas como la “silhouette score”).

---

###**Algoritmos más comunes de análisis de conglomerados:**

| Algoritmo                         | Tipo           | Descripción breve                                                              |
| --------------------------------- | -------------- | ------------------------------------------------------------------------------ |
| **K-Means**                       | Particional    | Divide los datos en *k* grupos minimizando la distancia interna de cada grupo. |
| **Hierarchical Clustering**       | Jerárquico     | Crea una jerarquía de grupos (árbol o dendrograma).                            |
| **DBSCAN**                        | Densidad       | Agrupa puntos densamente conectados y separa el ruido.                         |
| **Gaussian Mixture Models (GMM)** | Probabilístico | Asume que los datos provienen de una mezcla de distribuciones gaussianas.      |




---

###**Aplicaciones en IA y ciencia de datos:**

* Segmentación de clientes en marketing.

* Agrupamiento de documentos o noticias por tema.

* Reconocimiento de patrones (por ejemplo, en imágenes o comportamiento).

* Reducción de ruido o detección de anomalías.

* Exploración de datos antes de aplicar modelos supervisados.

##**AGRUPACION JERARQUICA AGLOMERATIVA**

El método jerarquico **es util para ordenar grupos en la estructura de jerarquía natural.**

Hay dos métodos de agrupamiento jerárquico, que son el agrupamiento aglomerativo y el agrupamiento divisivo:

**Agrupación aglomerativa**
* De abajo hacia arriba
* Comienza con n-clusters para aglomerar progresivamente clusters similares hasta que queda un cluster final.

**Agrupación divisiva**
* De arriba hacia abajo
* Comienza desde un solo grupo que incluye todos los registros y lo divide en n grupos (el agrupamiento aglomerativo es un método jerarquico utilizado principalmente).

---

###**AGLOMERATIVO**

**COMO FUNCIONA?**

El clustering jerárquico aglomerativo comienza con cada punto como su propio cluster y, en iteraciones sucesivas, une los dos clusters más “cercanos” según una regla de distancia entre clusters (linkage). Repite hasta que todos los puntos están en un único cluster. El resultado completo es una jerarquía de fusiones que se puede cortar a la altura deseada para obtener una partición.

###**QUE ES UN DENDOGAMA?**

Un dendrograma es un diagrama en forma de árbol que representa agrupaciones o relaciones jerárquicas entre datos, mostrando cómo los elementos se fusionan o dividen en niveles sucesivos.


image.png

image.png

###**QUE ES UNA REGLA DE ENLACE?**

1. En el análisis de conglomerados jerárquico aglomerativo, el algoritmo comienza considerando cada punto como su propio cluster.

2. Después, en cada paso, el algoritmo debe decidir qué dos clusters unir.

**Para eso necesita medir la distancia entre clusters, no solo entre puntos individuales.**

Esa distancia entre clusters no está definida de forma única, por eso existen distintas reglas de enlace (linkage) que determinan cómo calcular la distancia entre dos grupos.

---

**Supongamos que tenemos dos clusters:**

* Cluster A = {a₁, a₂, a₃}

* Cluster B = {b₁, b₂}

y queremos medir qué tan lejos están uno del otro.

Cada punto tiene una distancia hacia los puntos del otro cluster.

**Por ejemplo**, podemos calcular todas las distancias entre cada punto de A y cada punto de B:

    d(a₁, b₁), d(a₁, b₂), d(a₂, b₁), d(a₂, b₂), d(a₃, b₁), d(a₃, b₂)

Una vez que tenemos esas distancias, aplicamos una regla de enlace (linkage) que resume todas esas distancias en una sola medida de distancia entre A y B.

---

**TIPOS DE LINKAGE MAS COMUNES**

Existen varias formas de definir esa distancia.
Las más usadas son las siguientes:

image.png

a) **Single linkage (enlace sencillo o mínimo)**

La distancia entre dos clusters es la menor distancia entre cualquier par de puntos (uno de cada cluster).

image.png

**Interpretación:**
Une los clusters que tienen puntos más cercanos entre sí.

Tiende a formar clusters alargados o encadenados, porque basta que dos puntos estén cerca para unir grupos grandes.

**Ejemplo:**
Si el punto a₃ de A está muy cerca del punto b₁ de B, se unirán los clusters A y B aunque el resto de puntos estén más lejos.

---


b) **Complete linkage (enlace completo o máximo)**

La distancia entre dos clusters es la mayor distancia entre cualquier par de puntos (uno de cada cluster).

image.png

**Interpretación:** Solo une clusters cuando todos sus puntos están relativamente cerca.
Genera clusters compactos y de forma más redondeada.

Evita los efectos encadenados del single linkage.

**Ejemplo:**
Si un solo punto entre dos clusters está muy alejado, eso retrasa su unión.

---


c) **Average linkage (enlace promedio)**

La distancia entre dos clusters es el promedio de todas las distancias entre pares de puntos (uno de cada cluster).

image.png


**Interpretación:**
Toma una decisión intermedia entre las dos anteriores.

Forma clusters de tamaño más equilibrado y evita que los outliers afecten demasiado.

---


d) **Centroid linkage (enlace por centroide)**

Cada cluster tiene un centroide (el promedio de las coordenadas de sus puntos).

La distancia entre dos clusters es la distancia entre sus centroides.

image.png


**Interpretación:**
Rápido de calcular y fácil de entender, pero puede tener un problema: el fenómeno de inversión.

Es decir, al unir clusters, a veces la distancia a un tercer cluster puede disminuir, lo que rompe la estructura estrictamente jerárquica.

---


e) **Ward linkage (método de Ward)**

No mide distancias directamente, sino el aumento de la varianza interna al unir dos clusters.
Elige unir los clusters cuya fusión incrementa menos la varianza total dentro de los grupos.

image.png

**Interpretación:**
Tiende a formar clusters compactos y de tamaño similar.

Es muy utilizado en práctica porque suele producir resultados estables y fácilmente interpretables.

---

###**COMO AFECTA EL LINKAGE AL RESULTADO?**

El tipo de linkage determina cómo se unen los grupos y, por tanto, la forma final de los clusters.

**Por ejemplo:**

| Tipo de enlace | Forma de los clusters   | Ventajas                                  | Desventajas                                     |
| -------------- | ----------------------- | ----------------------------------------- | ----------------------------------------------- |
| Single         | Alargados o encadenados | Detecta estructuras alargadas             | Sensible al ruido y puntos atípicos             |
| Complete       | Compactos               | Grupos más definidos                      | Puede romper clusters naturales grandes         |
| Average        | Intermedios             | Balance entre sensibilidad y compactación | Costo computacional algo mayor                  |
| Centroid       | Esféricos               | Simple y rápido                           | Posibles inversiones                            |
| Ward           | Esféricos y balanceados | Muy usado, buena separación               | Supone que los clusters son similares en tamaño |





---

###**COMO SE USA DENTRO DEL ALGORITMO JERARQUICO?**

1. Se empieza con cada punto como un cluster.

2. Se calculan todas las distancias entre clusters (según el tipo de linkage elegido).

3. Se unen los dos clusters más cercanos.

4. Se actualizan las distancias entre los nuevos clusters y los demás.

5. Se repite el proceso hasta que quede solo un cluster.


**Cada fusión puede representarse en un dendrograma**, donde la altura de la unión refleja la distancia o el nivel de similitud entre los clusters que se están uniendo.

---

###**EJEMPLO CONCEPTUAL**

Imaginemos que tenemos cuatro puntos:

* A y B muy cercanos entre sí.

* C algo más lejos.

* D mucho más lejos.

.

**Con single linkage**, A y B se unen primero; luego C se une rápido a ese grupo porque está cerca de alguno de sus puntos.

**Con complete linkage**, A y B se unen primero, pero C no se une hasta que está razonablemente cerca de todos los puntos del cluster AB.

**Con Ward linkage**, A y B se unen si su fusión no aumenta mucho la varianza; C se une después si su adición mantiene el grupo compacto.

**EJERCICIO EN PYTHON**

```python
# Importamos las librerías necesarias
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import silhouette_score

from scipy.cluster.hierarchy import dendrogram, linkage # Libreria para el Dendograma

# 1. Cargamos el dataset Iris
iris = load_iris()
X = iris.data

print("Datos cargados con forma:", X.shape)

# 2. Escalamos los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Creamos el modelo de clustering jerárquico aglomerativo
model = AgglomerativeClustering(
    n_clusters=3,        # número de clusters finales deseados
    linkage='average',      # método de fusión (Ward minimiza la varianza dentro de los clusters)
    metric='euclidean'   # nueva forma de especificar la distancia (reemplaza affinity)
)

# 4. Ajustamos el modelo a los datos escalados - hacer predicciones
labels = model.fit_predict(X_scaled)

print(labels)

print('-'*80)

print('Siluette Score =', silhouette_score(X_scaled, labels))

print('-'*80)

# 5. Visualizamos los resultados
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis')
plt.title("Clustering Jerárquico Aglomerativo (Dataset Iris)")
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.show()

#Dendograma
respuesta = linkage(X_scaled, method='average')

plt.figure(figsize=(12,6))
plt.title('Dendrograma')
plt.xlabel('Muestras/Instancias')
plt.ylabel('Distancia')
dendrograma = dendrogram(respuesta)
plt.show()
```

**Opciones de Parametros de Linkage:**

| Valor        | Cómo funciona                                                         | Cuándo usarlo                                                                               |
| ------------ | --------------------------------------------------------------------- | ------------------------------------------------------------------------------------------- |
| `'ward'`     | Minimiza la varianza interna. Solo acepta métrica **euclidiana**.     | Cuando se espera que los clusters sean **esféricos o compactos**.                           |
| `'complete'` | Usa la **mayor distancia** entre puntos de dos clusters.              | Si quieres **clusters bien separados**, incluso si son de forma irregular.                  |
| `'average'`  | Usa la **distancia promedio** entre todos los puntos de dos clusters. | Si tus datos tienen **ruido o clusters de distintos tamaños**.                              |
| `'single'`   | Usa la **menor distancia** entre puntos de dos clusters.              | Tiende a producir “**efecto cadena**” (clusters alargados). Útil para detectar continuidad. |


---

**El Silhouette Score mide qué tan bien está asignado cada punto a su cluster.
Valores posibles:**

* +1: punto muy bien asignado

* 0: punto en el límite entre clusters

* -1: punto mal asignado (debería estar en otro grupo)

##**AGRUPACION JERARQUICA DIVISIVA**

El agrupamiento jerárquico divisivo es un método de análisis de conglomerados (o clustering) que comienza con todos los datos formando un solo grupo y los divide de forma progresiva hasta que cada observación queda en su propio cluster o hasta alcanzar un número deseado de grupos.

Por eso se lo llama “de arriba hacia abajo” (top–down), en contraste con el método aglomerativo, que empieza con muchos clusters individuales y los va uniendo.

image.png

###**COMO TRABAJA?**

**El método parte de un solo grupo con todos los puntos. En cada paso:**

1. Selecciona un cluster existente (por ejemplo, el más grande o el menos homogéneo).

2. Busca una forma de dividir ese cluster en dos subgrupos que estén lo más separados posible entre sí.

3. Repite el proceso con los nuevos grupos hasta llegar al número de clusters deseado o hasta que cada punto esté solo.

.

**Visualmente, el dendrograma del método divisivo se construye de arriba hacia abajo, comenzando con una raíz que representa todos los datos y ramas que se dividen sucesivamente.**

###**FUNCIONAMIENTO PASO A PASO**

Supongamos que tenemos un conjunto de datos con varios puntos:

* **Paso 1: Un único cluster inicial**

    Todos los puntos pertenecen a un solo grupo grande. Se calcula la medida de disimilitud interna (por ejemplo, la varianza o distancia promedio entre puntos).

* **Paso 2: Elegir un criterio de división**

    Se identifica el cluster más “heterogéneo” (el que tiene mayor dispersión interna).
    Ese cluster será dividido.

* **Paso 3: Separación del cluster**

    Para dividirlo, se puede usar un criterio de distancia o una técnica auxiliar, como:

    * Separar los puntos que están más alejados del centroide del cluster.

    * Aplicar un método de partición dentro de ese cluster

    * Usar medidas de similitud para encontrar los dos subconjuntos más distintos.

* **Paso 4: Actualizar la estructura jerárquica**

    La división produce dos clusters hijos.

    Cada uno se analiza nuevamente para ver si necesita dividirse más y en cada división, se actualiza el dendrograma, que muestra ramas que se van separando hacia abajo.

* **Paso 5: Repetir**

    El proceso se repite hasta:

    * Que cada punto esté en su propio cluster. * Que se alcance un número deseado de grupos (estos son definido por el analista).

---

###**EJEMPLO CONCEPTUAL**

Supongamos que tenemos los puntos

    A, B, C, D, E, F, G, H, I.

---

**Paso 1:**

Comenzamos con un solo cluster con todos los puntos:

      {A, B, C, D, E, F, G, H, I}



**Paso 2:**

Medimos la disimilitud interna.
Detectamos que el grupo tiene dos subconjuntos claros:

      {A, B, C, D} y {E, F, G, H, I}

Se realiza la primera división.
.

**Paso 3:**

Analizamos cada nuevo grupo por separado:

El grupo

      {A, B, C, D}

es relativamente homogéneo, no se divide.

Mientras que el grupo

      {E, F, G, H, I}

muestra dos subgrupos naturales:

      {E, F, G} y {H, I}

Se realiza la segunda división.



**Paso 4:**

Continuamos hasta que cada grupo sea internamente homogéneo o se alcance el número de clusters deseado.

El dendrograma resultante mostrará una estructura de árbol invertida, donde la raíz representa todos los datos y las ramas representan divisiones sucesivas.

---

**EJEMPLO EN PYTHON**

```python
# 1. Importamos las librerías necesarias
from sklearn.datasets import load_iris
from sklearn.cluster import BisectingKMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import silhouette_score

# 2. Cargamos el dataset Iris
iris = load_iris()
X = iris.data
y = iris.target

df = pd.DataFrame(X, columns=iris.feature_names)
print("Primeras filas del dataset:")
print(df.head())
print('-'*80)

# 3. Escalamos los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. Creamos el modelo de clustering jerárquico divisivo
model = BisectingKMeans(
    n_clusters=3,        # número final de clusters que queremos
    init='k-means++',    # inicialización inteligente de centroides
    n_init=10,           # número de veces que se repite con distintas inicializaciones
    max_iter=300,        # máximo de iteraciones por división
    random_state=42      # semilla para reproducibilidad
)

# 5. Ajustamos el modelo al dataset escalado - predicciones
labels = model.fit_predict(X_scaled)


# 6. Visualizamos los resultados
plt.figure(figsize=(7, 5))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='plasma')
plt.title("Clustering Jerárquico Divisivo (Bisecting K-Means) - Dataset Iris")
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.show()

print('-'*80)

# 7. Análisis simple del resultado
df['cluster'] = labels
print("\nCantidad de elementos en cada cluster:")
print(df['cluster'].value_counts())

print('-'*80)

print('Silhouette Score:', silhouette_score(X_scaled, labels))

```

##**DIFERENCIAS ENTRE LOS METODOS**

###**DIFERENCIAS ENTRE ESTOS**

| Característica      | Jerárquico aglomerativo                       | Jerárquico divisivo                           |
| ------------------- | --------------------------------------------- | --------------------------------------------- |
| Dirección           | De abajo hacia arriba                         | De arriba hacia abajo                         |
| Inicio              | Cada punto es un cluster                      | Todos los puntos forman un solo cluster       |
| Operación principal | Fusión de clusters                            | División de clusters                          |
| Dendrograma         | Se construye uniendo ramas hacia arriba       | Se construye dividiendo ramas hacia abajo     |
| Computación         | Más eficiente (O(n²))                         | Más costoso (puede ser O(2ⁿ))                 |
| Uso habitual        | Más común en práctica                         | Menos usado, pero útil en grandes estructuras |
| Ejemplo visual      | Óvalos que se unen (como la imagen que viste) | Óvalos grandes que se van separando           |


###**Guía práctica**


| Situación                                            | Método recomendado                   | Razón                                           |
| ---------------------------------------------------- | ------------------------------------ | ----------------------------------------------- |
| No sabes cuántos clusters hay                        | Aglomerativo                         | Permite decidirlo visualmente en el dendrograma |
| Quieres explorar relaciones locales                  | Aglomerativo                         | Une puntos similares de forma progresiva        |
| Tienes grandes grupos claros y quieres subdividirlos | Divisivo                             | Separa de lo general a lo específico            |
| Datos muy numerosos o heterogéneos                   | Divisivo (con k-means o heurísticas) | Escalable y más controlado                      |
| Necesitas clusters compactos y equilibrados          | Ward (en aglomerativo o divisivo)    | Minimiza la varianza interna                    |
| Buscas una jerarquía conceptual o lógica             | Divisivo                             | Más intuitivo para clasificaciones jerárquicas  |


###**Cuándo conviene usar el método jerárquico aglomerativo?**

Conviene cuando...

> **Cuando el conjunto de datos es pequeño o mediano**

* El aglomerativo requiere calcular todas las distancias entre los puntos (complejidad cuadrática: O(n²)).

* Para conjuntos de tamaño pequeño o moderado (hasta unos pocos miles de observaciones) es perfectamente viable.

* En cambio, el divisivo puede volverse exponencialmente costoso.

**Ejemplo:**
Un análisis de 50 o 200 clientes para segmentar perfiles de compra, Aqui el aglomerativo es ideal.

.

> **Cuando se desconoce la cantidad de clusters**

* El método jerárquico genera un dendrograma completo que muestra todas las posibles fusiones.

* Luego se puede decidir cuántos clusters mantener simplemente “cortando” el dendrograma a una cierta altura.

**Ejemplo:**
Estudio exploratorio de un conjunto de plantas o productos químicos sin saber cuántos grupos naturales existen.

.

> **Cuando los clusters pueden tener tamaños o formas diferentes**

Según la regla de enlace elegida (linkage), el aglomerativo puede adaptarse a distintas estructuras de datos.

* Single linkage capta grupos alargados.

* Complete linkage da clusters compactos.

* Ward produce clusters equilibrados.

**Ejemplo:**
Datos biológicos, ecológicos o de comportamiento, donde los grupos no son uniformes.

> **Cuando se necesita una representación jerárquica completa**

* El dendrograma del método aglomerativo muestra de manera visual y progresiva cómo se van uniendo los grupos.

* Es útil en análisis descriptivo, minería de datos y bioinformática (por ejemplo, árboles filogenéticos).

Ejemplo:
Agrupar genes o especies por similitud genética.

###**Ejemplo de razonamiento aplicado**

**Supongamos que estás analizando los datos de comportamiento de clientes de un banco...**

> Si no sabes cuántos tipos de clientes hay y quieres explorar patrones, el aglomerativo es ideal: te da un dendrograma completo que puedes cortar donde mejor se interpreten los grupos.

> Si ya sabes que existen grandes categorías (por ejemplo, clientes personales, empresariales, e institucionales), pero quieres subdividirlas según su nivel de actividad, el divisivo es más natural: partes del grupo global y lo vas separando en ramas.

---



###**Cuándo conviene usar el método jerárquico divisivo?**

Conviene cuando...

> **Cuando el conjunto de datos es grande o muy heterogéneo**

* El divisivo es más eficiente si se aplica con estrategias heurísticas (como usar k-means para dividir).

* Puede evitar los errores de fusión temprana del método aglomerativo, ya que parte de una estructura global y la va refinando.

**Ejemplo:**
Análisis de documentos o clientes donde hay un gran grupo inicial y se busca segmentarlo progresivamente.

> **Cuando hay una estructura general clara y queremos refinarla**

* Si sabemos que los datos pertenecen a un grupo amplio, pero queremos ver las subcategorías internas, el divisivo es más natural.

* Por ejemplo, primero separar “clientes activos” y “clientes inactivos”, y luego dividir cada grupo en segmentos más específicos.

**Ejemplo:**
Segmentación de mercado o jerarquías organizacionales.

> **Cuando interesa preservar la estructura global antes de separar**

* El divisivo no sufre tanto el problema de “errores tempranos” que tiene el aglomerativo (cuando une puntos que no deberían unirse y ya no puede deshacerlo).

* En el divisivo, las divisiones se controlan mejor porque siempre partimos del conjunto completo.

**Ejemplo:**
Clasificación de documentos o textos, donde primero se identifican grandes temas (Ciencia, Economía, Arte) y luego se subdividen en categorías más finas.

> **Cuando se usan criterios de homogeneidad global (como Ward)**

* Si se dispone de un criterio de varianza o cohesión bien definido (como el método de Ward), el divisivo puede producir resultados muy equilibrados.

* Esto es útil en estudios donde se necesita homogeneidad interna en cada grupo.

**Ejemplo:**
Agrupar zonas geográficas según indicadores socioeconómicos.
