###**쯈u칠 es un entorno virtual (venv)?**

Un entorno virtual es un espacio aislado donde puedes instalar librer칤as de Python sin afectar las del sistema.

Sirve para tener proyectos con dependencias distintas (por ejemplo, un proyecto usa streamlit y otro no).

**Con esto verificamos si tenemos venv instalado**

    python -m venv --help


**Para crear el entorno en la terminal, escribimos:**

    python -m venv nombre_del_entorno


**Por ejemplo:**

    python -m venv venv_streamlit


Esto crea una carpeta dentro del proyecto llamada venv_streamlit, que contiene Python y pip independientes.

**Para activarlo depende del sistema:**

**En Windows (PowerShell o CMD):**

    venv_streamlit\Scripts\activate

**En Linux o macOS:**

    source venv_streamlit/bin/activate


**Veran que su terminal cambia y ahora muestra algo as칤:**

    (venv_streamlit) PS C:\Users\Cleidys\mi_proyecto>


**Para instalar librerias**

    pip install streamlit pandas scikit-learn

###**Solucion de Error en Windows**

**Permitir la ejecuci칩n de scripts en PowerShell**

PowerShell, por razones de seguridad, bloquea la ejecuci칩n de scripts por defecto, incluyendo los que activan entornos virtuales.
Para permitirlo, sigue estos pasos:

Abran PowerShell como administrador y escriban el siguiente comando:

    Set-ExecutionPolicy RemoteSigned


Presionan Enter y luego escriban 'Y' para confirmar el cambio.

Este ajuste permite ejecutar scripts locales.

.

**Activamos de nuevo el entorno virtual**

Una vez hayas cambiado la pol칤tica de ejecuci칩n, puedes activar el entorno escribiendo en PowerShell:


    venv_streamlit\Scripts\activate
