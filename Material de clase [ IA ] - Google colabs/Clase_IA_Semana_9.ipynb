**Indexacion booleana**

* Es una forma de filtrar datos de un arreglo usando condiciones lógicas (verdadero o falso).

* En vez de recorrer todo el arreglo con bucles, NumPy usa un “filtro” de valores booleanos (True o False) que indica qué elementos cumplen la condición.

```python
import numpy as np

arr = np.array([5, 10, 15, 20, 25])

# Creamos un filtro
filtro = arr > 10

print("Filtro booleano:", filtro)
print("Elementos mayores que 10:", arr[filtro])
```

Porque usariamos esto?

#**PARTE 2 - Preprocesamiento y Visualizacion de Datos**

El preprocesamiento de datos es el conjunto de técnicas y pasos que se aplican a los datos crudos para dejarlos listos y adecuados para ser usados en un modelo de inteligencia artificial.

En otras palabras: **antes de entrenar un modelo, los datos no se pueden usar tal cual**, porque suelen tener problemas como:

* Valores faltantes o nulos

* Datos inconsistentes o ruidosos

* Diferentes escalas o unidades

* Datos categóricos que no se pueden interpretar directamente

El preprocesamiento soluciona estos problemas, **asegurando que el modelo aprenda correctamente** y dé resultados precisos.



---



**Etapas del preprocesamiento:**

1. Comprobar valores faltantes.

2. Reemplazar o eliminar datos erróneos.

3. Tratar duplicados.

4. Convertir tipos de datos.

5. Escalar y normalizar valores.

6. Generar nuevas características (feature engineering).

```python
import pandas as pd
import numpy as np

data = {
    'Nombre':['Ana', 'Luis', 'Pedro', 'Juan', 'Carla', 'Sofia', 'Sofia', 'Daniel'],
    'Edad':[20, np.nan, 22, 21, 23, 22.0, 22.0, 21],
    'Promedio':[85, 90, np.nan, 88, 92, 76,76, 92],
    'Carrera':['IA', np.nan, 'IA', 'Programacion', 'Cybersecurity', 'IA', 'IA', 'Programacion']
}

df = pd.DataFrame(data)
df
```

#**Extra: Que es KDE y como se usa**

En **Inteligencia Artificial**, cada columna de un dataset (cada feature o característica) puede tener una distribución distinta:

* Algunas son simétricas (parecen una campana, distribución normal)

* Otras son asimétricas (muchos valores pequeños y pocos grandes, distribución sesgada)

* Algunas tienen outliers (valores extremos)

**Y esto porque nos importa?**

1. Porque influe para la normalización y transformación

2. Muchos algoritmos de machine learning funcionan mejor si las características tienen ciertas propiedades.





| Tipo de algoritmo                   | Preferencia de distribución                                      |
| ----------------------------------- | ---------------------------------------------------------------- |
| Regresión lineal / redes neuronales | Normalizada o estándar (media=0, desviación=1)                   |
| Árboles de decisión / Random Forest | Menos sensible, no requiere normalización                        |
| PCA (reducción de dimensiones)      | Datos centrados y escalados para capturar varianza correctamente |


Entonces, si vemos que alguna de nuestras caracteristicas está muy sesgada o tiene valores muy dispersos, tenemos estas opciones para resolverlo:

* Normalización (MinMaxScaler) = Llevar todos los valores entre 0 y 1

* Estandarización (StandardScaler) = Media=0, desviación=1

* Transformaciones logarítmicas o raíz cuadrada = Para reducir asimetría o suavizar outliers

**Como ayuda KDE?**

Este dibuja una curva suave de densidad y ...

1. Si la curva parece asimétrica o con colas largas, sabes que necesitas transformar la variable.

2. Si la curva es aproximadamente normal, probablemente no necesitas normalización compleja.

**Ejemplo de referencia:**

```python
# Importamos librerías
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Configuración de Seaborn
sns.set_style("whitegrid")

# Creamos datos
np.random.seed(42)

# Variable normal
edad = np.random.normal(loc=30, scale=5, size=1000)

# Variable asimétrica
ingreso = np.random.exponential(scale=2000, size=1000)


# Graficamos las distribuciones originales

plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
sns.histplot(edad, kde=True, bins=30, color="skyblue")
plt.title("Edad (aproximadamente normal)")

plt.subplot(1,2,2)
sns.histplot(ingreso, kde=True, bins=30, color="salmon")
plt.title("Ingreso (asimétrica con colas largas)")

plt.show()

# Aplicamos transformación logarítmica

ingreso_log = np.log1p(ingreso)  # log(1 + x) para evitar log(0)

plt.figure(figsize=(7,5))
sns.histplot(ingreso_log, kde=True, bins=30, color="green")
plt.title("Ingreso transformado con log (distribución más normal)")
plt.show()
```
