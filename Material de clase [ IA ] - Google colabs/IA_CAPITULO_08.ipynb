###**¿Qué es una API?**

**Una API (Application Programming Interface)** es un conjunto de funciones, métodos o reglas que permiten que un software interactúe con otro.
En el contexto de frameworks de Machine Learning:

* Una API define cómo usas las funciones y clases del framework.

* Por ejemplo, la API de Keras define cómo crear modelos, capas, entrenarlos, etc.

---

##**¿Qué es Keras?**

Keras es una API de alto nivel especializada en redes neuronales que funciona sobre distintos "motores" de computación (TensorFlow, JAX, PyTorch —aunque el soporte varía).
En la práctica moderna, Keras es la API oficial incluida en TensorFlow, ubicada en:

```python
tf.keras
```

**Keras no es un framework completo por sí solo:**

* es una capa de abstracción que simplifica tareas que, usando únicamente TensorFlow puro, serían demasiado bajas a nivel matemático o muy verbosas.

###**¿Por qué se creó Keras?**

**Antes del 2017:**

* TensorFlow puro era difícil de usar.

* La programación requería construir grafos manualmente.

* El código era largo, complejo y difícil de depurar.

Keras apareció como un interfaz amigable, inspirada en la filosofía:

> **“La simplicidad acelera la innovación.”**

Desde TensorFlow 2.0, Keras se integró nativamente, convirtiéndose en su forma principal de uso.

.

---

###**¿Cuál es la función principal de Keras?**

Hacer que diseñar y entrenar modelos de IA sea rápido, fácil y legible, sin perder la flexibilidad del backend de TensorFlow.

**Por ejemplo:**

* Crear redes densas

* Trabajar con CNN, RNN, Transformers

* Entrenar modelos distribuidos en GPU/TPU

* Usar callbacks como EarlyStopping

* Hacer tuning de hiperparámetros

* Guardar y exportar modelos

Keras proporciona estas herramientas en una interfaz clara y compacta.

###**Arquitectura conceptual de Keras**

**Keras está construida en torno a tres grandes conceptos:**


###**1. Modelos**

Son estructuras principales que definen la topología de la red.

**Tipos:**

* **Sequential:** capas lineales una tras otra

* **Functional API:** modelos complejos con múltiples entradas/salidas

* **Subclassing:** modelos personalizados usando clases Python


###**2. Capas (Layers)**

**Keras provee más de 100 tipos de capas listas para usar:**

* **Dense:** Capa totalmente conectada; se usa para clasificación/regresión clásica. Es la capa más común y versátil.

* **Conv2D:** Capa de convolución para imágenes; detecta patrones visuales como bordes, texturas y objetos. Es esencial en visión por computadora.

* **LSTM:** Tipo de red recurrente que recuerda secuencias largas; ideal para texto, series temporales y procesamiento secuencial.

* **Embedding:** Convierte palabras/IDs en vectores densos; transforma categorías o texto en representaciones numéricas aprendibles.

* **Dropout:** Apaga aleatoriamente neuronas durante el entrenamiento; evita sobreajuste y mejora la generalización.

* **BatchNormalization:** Normaliza activaciones dentro del modelo; acelera el entrenamiento y estabiliza el aprendizaje.

* **Attention:** Mecanismo que aprende qué partes de la entrada son más importantes; base de Transformers y modelos modernos de NLP.

.

**Una capa en Keras encapsula:**

* pesos

* función de activación

* cálculo del forward pass

###**3. Entrenamiento**

**Keras ofrece una API estandarizada para entrenar modelos:**

####**¿Qué es un “tensor”?**

Un tensor es una estructura de datos que representa números en múltiples dimensiones.

**Es la generalización de:**

* escalar → (0D)

* vector → (1D)

* matriz → (2D)

* cubo de datos → (3D)

* volúmenes n-dimensionales → (nD)

.

**Tensor = contenedor de números + forma (shape) + tipo de dato (dtype).**

image.png

**TensorFlow llama todo “tensor” porque opera con datos en múltiples dimensiones.**

.

**Porque TensorFlow funciona haciendo fluir tensores a través de un grafo computacional.**

* Los tensores son los datos.

* El flow es el conjunto de operaciones matemáticas que se aplican a ellos.

image.png

###**¿Qué hace TensorFlow con los tensores?**

**Cuando entrenas un modelo:**

1. Recibe tensores como entrada (por ejemplo, imágenes o vectores).

2. Pasa los tensores por capas (matrices de pesos).

3. Ejecuta operaciones matemáticas (multiplicaciones, convoluciones, activaciones).

4. Computa la pérdida.

5. Calcula gradientes automáticamente (autodifferentiation).

6. Actualiza los pesos usando un optimizador.

7. Repite el ciclo hasta que el modelo aprende.

.

**TensorFlow todo lo hace a través de tensores.**

###**Y Ahora...**

Conociendo lo anterior, podemos imaginar un **tensor** como un **contenedor matemático de datos, similar a una hoja de cálculo** que **puede tener una celda (escalar), una columna (vector), una tabla (matriz) o un conjunto de tablas apiladas (tensor de orden superior).**

> **La diferencia es que este “contenedor” puede viajar entre dispositivos (CPU–GPU) sin cambiar de forma.**

####**Como creamos un Tensor?**

**Manualmente**

```python
import tensorflow as tf

# Escalar
a = tf.constant(5)

# Vector
b = tf.constant([1, 2, 3])

# Matriz
c = tf.constant([[1, 2], [3, 4]])

# Tensor de 3 dimensiones
d = tf.constant([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

print(a)
```

**Tensores desde NumPy**

```python
import numpy as np

arr = np.array([[10, 20, 30]])
tensor = tf.convert_to_tensor(arr)

print(tensor)
```

**Tensores llenos de ceros o unos**

```python
tf.zeros([3, 3])     # Matriz 3x3 llena de ceros
tf.ones([2, 5])      # Matriz 2x5 llena de unos
```

**Tensor aleatorio**

```python
tf.random.normal([4, 4], mean=0, stddev=1)
```

####**Operaciones básicas**

**1) Tensor shape (forma)**


```python
tensor.shape
```

**2) Cambiar forma (reshape)**

```python
x = tf.constant([[1,2,3,4,5,6]])
print(x)
print('-'*20)
y = tf.reshape(x, [6,1])   # de 1x6 a 2x3 o 6x1 = 6
print(y)
```

**3) Indexación**

```python
x = tf.constant([[10,20,30], [40,50,60]])
x[0]        # primer vector
x[:, 1]     # segunda columna
```

**4) Operaciones matemáticas**

```python
x = tf.constant([1,2,3])
y = tf.constant([10,20,30])

q = x + y
w = x * y

print(q)
print(w)

tf.matmul([[1,2]], [[3],[4]])   # multiplicación matrices
```

**5) Casting (cambiar tipo)**

```python
tf.cast(tf.constant([1,2,3], dtype=tf.int32), tf.float32)
```

**6) GradientTape para derivadas automáticas**

```python
x = tf.Variable(3.0)

with tf.GradientTape() as tape:
    y = x * x + 2*x + 5

dy_dx = tape.gradient(y, x)
```
