# **MINERIA DE TEXTOS**

La minería de textos (text mining) es el proceso de analizar grandes cantidades de texto no estructurado para extraer información útil, patrones, tendencias o conocimiento que no es evidente a simple vista.

**En otras palabras:**
- Toma textos,
- los procesa,
- extrae datos relevantes,
- y los convierte en información útil para decisiones o análisis.




###**¿Qué hace exactamente la minería de textos?**

**Incluye técnicas como:**

1. Limpieza y preparación del texto
(eliminar ruido, normalizar palabras, quitar stopwords)

2. Clasificación de documentos
(por ejemplo, identificar si un comentario es positivo o negativo)

3. Agrupamiento (clustering)
(agrupar textos similares, como noticias por tema)

4. Extracción de palabras clave y frases importantes

5. Detección de temas (topic modeling)

6. Análisis de sentimientos
(determinar emociones en textos)

7. Reconocimiento de entidades
(nombres de personas, lugares, organizaciones)


###**¿Para qué sirve?**

* Detectar fraudes o amenazas en ciberseguridad

* Analizar opiniones de clientes

* Automatizar clasificación de correos o documentos

* Estudiar tendencias en redes sociales

* Procesar grandes volúmenes de noticias o reportes

* Apoyo en investigaciones científicas o legales


### **Ejemplo:**

* Si tienes 10,000 comentarios de clientes, la minería de textos permite descubrir:

* cuáles son las palabras más frecuentes,

* si la mayoría está satisfecha o no,

* qué problemas se repiten,

* qué temas aparecen con más fuerza.

## **QUE SON LOS DATOS NO ESTRUCTURAADOS**

Los datos no estructurados son aquellos que no siguen un formato rígido, como textos libres (correos, mensajes, comentarios, artículos, PDFs, chats, etc.).
Para poder analizarlos, hay que transformarlos en algo que las máquinas sí puedan procesar.

---

**Datos que aun no estan estructurados. No tiene un modelo de datos definido (estructura)**

- Documentos, vídeos o audios que tienen una gran cantidad de datos pero con estructuras y formas variadas.

- Los libros, las revistas, los documentos, los metadatos, los historiales medicos, el audio, el vídeo, los datos analogicos, las imagenes, los archivos y tambien los mensajes de correo electronico, las paginas web y los documentos de los procesadores de texto estan compuestos por textos no estructurados.

## **Analisis de datos no estructurados**

### **1. Introduzca manualmente etiquetas en los metadatos para estructurar los textos**

**¿Qué significa?**  
Antes de analizar un texto, a veces es necesario ponerle etiquetas o categorías para indicar de qué trata o qué tipo de información contiene. Esto ayuda a que las máquinas puedan agrupar o clasificar textos correctamente.

**Ejemplos:**

* A un conjunto de correos se les puede poner etiquetas como "spam", "trabajo", "personal".
* A comentarios de clientes se les puede poner etiquetas como "positivo", "negativo", "neutro".
* A documentos legales se les puede etiquetar como "contrato", "sentencia", "factura".

**¿Por qué se hace?**  
Porque los textos por sí solos no tienen una estructura clara, y etiquetarlos facilita el análisis automático.

---

### **2. La estructuración de datos basada en la minería de textos utiliza un método que crea una etiqueta para que una palabra del texto y una parte del discurso se correspondan**

**¿Qué significa esto?**  
Aquí se habla de procesamiento del lenguaje natural (NLP). Existen técnicas que identifican automáticamente:

* Qué tipo de palabra es (verbo, sustantivo, adjetivo).
* Qué función tiene dentro de la oración.
* Qué entidades aparecen (personas, lugares, organizaciones).

El sistema etiqueta cada palabra según su rol lingüístico.

**Ejemplos:**

En la frase:  
"El servidor falló por una sobrecarga inesperada."

Las etiquetas serían:

* El → determinante  
* servidor → sustantivo  
* falló → verbo  
* sobrecarga → sustantivo  
* inesperada → adjetivo  

Esto ayuda a los algoritmos a entender qué ocurre en la oración.

**En minería de textos esto se usa para:**

* análisis de sentimientos  
* clasificación automática  
* encontrar relaciones entre palabras  
* detectar patrones lingüísticos

---

### **3. Utiliza un software que construye estructuras procesables por máquinas**

**¿Qué significa?**  
El texto se transforma en representaciones numéricas que las computadoras pueden interpretar. Las máquinas no entienden palabras, sino valores matemáticos.

Esto puede hacerse con técnicas como:

* Bag of Words  
* TF-IDF  
* Word embeddings  
* Transformers (como BERT)

**Ejemplo:**  
La palabra "seguridad" puede convertirse en un vector como:

[0.42, 0.15, -0.37, 0.88, ...]

Ese vector representa su significado en relación con miles de textos.

---

### **4. Análisis que permite la deducción semántica a partir de textos, sintaxis y otros patrones pequeños o grandes**

**¿Qué significa?**  
Se refiere a la capacidad del sistema de comprender el significado del texto, no solo palabras aisladas. La máquina analiza:

* Sintaxis: cómo se estructura la oración.  
* Semántica: qué quiere decir realmente.  
* Patrones: tendencias repetidas.

**Ejemplos:**

* Análisis de sentimientos: determina si un comentario es positivo o negativo.  
* Detección de temas: en un conjunto de noticias puede descubrir temas como economía, salud, deportes o política.  
* Identificación de spam o phishing en ciberseguridad: detecta patrones comunes de correos fraudulentos.  
* Detección de amenazas en redes sociales: reconoce frases como "voy a dañar", "tengo un arma", "ataque planificado".

###**Mineria de datos Vs Mineria de textos**

La minería de datos extrae patrones utiles y valiosos de los datos estructurados.

En cambio, la minería de textos extrae entidades con nombre, patrones o informacion sobre relaciones entre
palabras y frases a partir de datos no estructurados compuestos por lenguaje natural

# **Técnicas de Minería de Textos**

La minería de textos utiliza métodos que permiten transformar texto sin estructura en información útil. Estas técnicas provienen del procesamiento del lenguaje natural (NLP), la estadística y el aprendizaje automático.

A continuación se explican las más importantes.


## 1. Preprocesamiento del texto

Antes de analizar, el texto debe limpiarse y normalizarse.

Incluye pasos como:

* Eliminación de ruido (signos innecesarios, HTML, correos, URLs).
* Tokenización (dividir el texto en palabras o frases).
* Eliminación de *stopwords* (palabras sin valor analítico como "de", "la", "el").
* Lematización o *stemming* (reducir palabras a su forma base: comprando → comprar).
* Normalización (pasar todo a minúsculas, corregir caracteres).

**¿Para qué sirve?**  
Para dejar el texto listo para análisis y evitar distracciones del modelo con palabras irrelevantes.


## 2. Representación del texto

Como las máquinas no entienden palabras, deben convertirse en números.

Técnicas más usadas:

### a) Bag of Words (BOW)
* Cuenta cuántas veces aparece cada palabra.  
* Simple pero útil para tareas básicas.

### b) TF-IDF
* Mide la importancia de una palabra en un documento según cómo aparece en otros.  
* Identifica términos relevantes.

### c) Word Embeddings
* Convierte palabras en vectores que representan su significado.  
* Ejemplos: **Word2Vec**, **GloVe**.  
* Palabras similares → vectores similares.

### d) Modelos de lenguaje modernos (Transformers)
* Representan el texto con su contexto completo.  
* Ejemplos: **BERT**, **GPT**.  
* Son los más precisos hoy en día.


## 3. Clasificación de textos

Se asigna el texto a una categoría predefinida.

Ejemplos:

* Spam o no spam.
* Opiniones positivas, negativas o neutras.
* Tipos de documentos.

Algoritmos: Naive Bayes, SVM, redes neuronales, Transformers.


## 4. Análisis de sentimientos

Determina la emoción expresada en un texto.

Puede detectar:

* satisfacción
* frustración
* enojo
* alegría
* preocupación

Usado en marketing, atención al cliente y reputación.


## 5. Extracción de palabras clave

Identifica las palabras o frases más relevantes del texto.

Técnicas:

* TF-IDF
* RAKE
* KeyBERT
* N-grams

Sirve para resúmenes, motores de búsqueda y análisis rápido.


## 6. Reconocimiento de entidades (NER)

Detecta nombres propios dentro del texto.

Ejemplos:

* personas
* empresas
* ciudades
* fechas
* productos
* leyes
* ubicaciones

Útil en análisis legal, noticias, seguridad y minería empresarial.



## 7. Modelado de temas (Topic Modeling)

Descubre temas principales dentro de grandes colecciones de textos.

Algoritmos:

* LDA (Latent Dirichlet Allocation)
* NMF (Non-Negative Matrix Factorization)
* BERTopic

Ejemplo:  
En 10,000 tweets se identifican temas como política, fútbol y tecnología.



## 8. Clustering o agrupamiento

Agrupa textos similares sin etiquetas previas.

Ejemplo:  
Agrupar reseñas de clientes según los problemas mencionados.

Es útil cuando no se conoce la estructura del conjunto de datos.



## 9. Resumen automático de textos

Genera un resumen corto.

Tipos:

* **Extractivo:** selecciona frases importantes.  
* **Abstractivo:** redacta un resumen nuevo.

Aplicaciones: informes, artículos, documentos legales.



## 10. Detección de anomalías o contenido inusual

Encuentra textos que no encajan con el comportamiento normal.

**Ejemplos:**

* correos sospechosos (phishing)
* comentarios tóxicos
* mensajes extraños
* patrones de fraude



### Resumen

Las técnicas de minería de textos permiten:

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import nltk
from nltk.corpus import stopwords

nltk.download("stopwords")

# Corpus simple (colección de textos)
data = {
    "texto": [
        "Me encantó el producto, es excelente",
        "Muy malo, no funciona",
        "Lo recomiendo, buena calidad",
        "Pérdida de dinero, pésima experiencia",
        "Estoy feliz con la compra",
        "Terrible, llegó roto",
    ],
    "sentimiento": ["positivo", "negativo", "positivo", "negativo", "positivo", "negativo"]
}

df = pd.DataFrame(data)

X_train, X_test, y_train, y_test = train_test_split(
    df["texto"], df["sentimiento"], test_size=0.3, random_state=42
)

# TF-IDF convierte texto en vectores numéricos
stopwords_es = stopwords.words("spanish")
vectorizer = TfidfVectorizer(stop_words=stopwords_es)

X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Entrenamiento del modelo
modelo = MultinomialNB()
modelo.fit(X_train_tfidf, y_train)

# Evaluación
predicciones = modelo.predict(X_test_tfidf)
print("Exactitud:", accuracy_score(y_test, predicciones))

# Prueba con texto nuevo
nuevo = ["El producto llegó rápido y funciona perfecto"]
nuevo_vec = vectorizer.transform(nuevo)
print("Predicción:", modelo.predict(nuevo_vec)[0])
```

## **Explicación Teórica**
**¿Qué es un corpus?**

Un corpus es simplemente una colección de textos que se utiliza para entrenar o analizar en un proyecto de minería de textos o NLP.

**En este ejemplo, nuestro corpus es:**

* una lista pequeña de frases de clientes

* cada frase tiene una etiqueta: positivo o negativo

**En proyectos reales, el corpus puede ser:**

* miles de correos

* publicaciones de redes sociales

* opiniones de productos

* artículos de noticias

* documentos legales

**El corpus es la materia prima del análisis.**

## **¿Qué es TF-IDF y por qué se usa?**

TF-IDF significa:

* TF (Term Frequency): frecuencia de una palabra dentro de un documento

* IDF (Inverse Document Frequency): qué tan rara es esa palabra en el corpus

* Sirve para medir la importancia real de cada palabra.

**Ejemplo:**

“excelente” aparece poco → es importante

“el”, “de”, “muy” aparecen en todas partes → no son importantes

TF-IDF convierte cada texto en un vector numérico como:

[0.0, 0.54, 0.33, 0.0, ...]


**Cada número representa la importancia de una palabra.**

**Las máquinas no entienden español, pero sí entienden números.**
